# Segdups

Looking at SVs (INS and DEL) In segdups

I basically annotated the SVs as in_SD or not_in_SD and now i will plot stability profiles of G4s around INS that are in_SD and not_in_SD and same for DEL.

See if breakdown changes anything

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import fisher_exact

# ==== Load ====
df = pd.read_csv("/home/alextu/scratch/segdup_scripts/all_combined_quadron_results_SV_flanks_withSDs.csv")

# ==== Counts of G4s ====
counts = df.groupby("SD").size().to_dict()
count_in  = counts.get("in_SD", 0)
count_out = counts.get("not_in_SD", 0)

# ==== Bases of G4s (sum of L) ====
base_counts = df.groupby("SD")["L"].sum().to_dict()
g4_in  = base_counts.get("in_SD", 0)
g4_out = base_counts.get("not_in_SD", 0)

print("Counts -> in_SD:", count_in, "not_in_SD:", count_out)
print("Bases  -> in_SD:", g4_in,   "not_in_SD:", g4_out)

# ==== Fisher’s exact test ====
# Test enrichment of G4 bases in_SD vs not_in_SD
# If no total flank lengths available, a symmetric test is used:
table = [[g4_in, g4_out],
         [count_in, count_out]]  # using counts as complementary category

oddsratio, pvalue = fisher_exact(table)
print("Fisher’s exact test -> odds ratio:", oddsratio, "p-value:", pvalue)

# ==== Plotting ====
fig, axes = plt.subplots(1, 2, figsize=(10,4))

# Counts
pd.Series(counts).plot(kind="bar", ax=axes[0], color=["steelblue","orange"])
axes[0].set_ylabel("Number of G4 motifs")
axes[0].set_title("G4 counts in vs out SDs")

# Bases
pd.Series(base_counts).plot(kind="bar", ax=axes[1], color=["steelblue","orange"])
axes[1].set_ylabel("Total G4 bases (sum of L)")
axes[1].set_title("G4 bases in vs out SDs")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re

# ========= CONFIG =========
file_path = '/home/alextu/scratch/segdup_scripts/all_combined_quadron_results_SV_flanks_withSDs.csv'
STABLE_Q = 19  # Q > 19 => Stable
SD_LEVELS = ["in_SD", "not_in_SD"]  # enforce ordering in plots
SV_ORDER = ["INS", "DEL"]
FLANK_ORDER = ["pre_flank", "post_flank"]
# ==========================

# ---------- Load & derive ----------
df = pd.read_csv(file_path)
# basic cleaning
df = df.dropna(subset=['Q', 'L'])
df['POS'] = pd.to_numeric(df['POS'], errors='coerce')
df['L']   = pd.to_numeric(df['L'],   errors='coerce')
df = df.dropna(subset=['POS','L'])

# midpoint (center of motif window relative to flank coords)
df["Midpoint"] = (df["POS"] + (df["POS"] + df["L"])) / 2

# flank from Sample
def extract_flank(sample_name: str):
    m = re.search(r'_(pre|post)_flank(?:\.txt)?$', str(sample_name))
    return f"{m.group(1)}_flank" if m else "Unknown"

# SV type from Sample
def extract_svtype(sample_name: str):
    m = re.search(r'(?:^|[_\-.])(INS|DEL)(?:[_\-.]|$)', str(sample_name), flags=re.IGNORECASE)
    return m.group(1).upper() if m else "Unknown"

df["Flank"]  = df["Sample"].apply(extract_flank)
df["SVType"] = df["Sample"].apply(extract_svtype)

df = df[(df["Flank"].isin(FLANK_ORDER)) & (df["SVType"].isin(SV_ORDER))]

# normalize SD labels and drop NA if present
df["SD"] = df["SD"].map(lambda x: "in_SD" if str(x).lower()=="in_sd"
                                  else ("not_in_SD" if str(x).lower()=="not_in_sd" else "NA"))
df = df[df["SD"].isin(SD_LEVELS)].copy()

# stability label
df["Stability"] = np.where(df["Q"] > STABLE_Q, "Stable", "Unstable")

# ---------- Helpers ----------
def avg_profile_by_midpoint(frame):
    """Return mean Q per Midpoint for each Flank × SD."""
    out = (frame.groupby(["Flank", "SD", "Midpoint"], as_index=False)["Q"]
                 .mean()
                 .rename(columns={"Q":"Avg_Q"}))
    # sort categorical orders for consistent plotting
    out["Flank"] = pd.Categorical(out["Flank"], categories=FLANK_ORDER, ordered=True)
    out["SD"]    = pd.Categorical(out["SD"],    categories=SD_LEVELS,  ordered=True)
    return out.sort_values(["Flank","SD","Midpoint"])

def plot_profiles_for_svtype(df_in, svtype_label):
    sub = df_in[df_in["SVType"] == svtype_label]
    if sub.empty:
        print(f"[WARN] No rows for {svtype_label}; skipping.")
        return

    prof = avg_profile_by_midpoint(sub)

    fig, axs = plt.subplots(1, 2, figsize=(16, 6), sharey=True)
    fig.suptitle(f"{svtype_label} — Avg Q vs Midpoint (in_SD vs not_in_SD)", fontsize=18, fontweight='bold')

    for i, flank in enumerate(FLANK_ORDER):
        ax = axs[i]
        ax.set_title(f"{flank.replace('_',' ').title()}", fontsize=16)
        ax.set_xlabel("Midpoint", fontsize=14)
        if i == 0:
            ax.set_ylabel("Average Q", fontsize=14)
        ax.axhline(y=STABLE_Q, color='black', linestyle='--', linewidth=1)

        for sd in SD_LEVELS:
            cur = prof[(prof["Flank"]==flank) & (prof["SD"]==sd)]
            if cur.empty:
                continue
            ax.plot(cur["Midpoint"], cur["Avg_Q"],
                    marker='o', linestyle='-', linewidth=2, markersize=4,
                    label=sd)

        ax.grid(True, linestyle='--', alpha=0.4)
        ax.legend(title="SD", fontsize=11)

    plt.tight_layout()
    plt.show()

def bar_counts(ax, cats, counts, title, ylabel="Count"):
    x = np.arange(len(cats))
    ax.bar(x, counts, width=0.8)
    ax.set_xticks(x, cats, rotation=0)
    ax.set_title(title, fontsize=14)
    ax.set_ylabel(ylabel)
    for xi, c in zip(x, counts):
        ax.text(xi, c, str(int(c)), ha='center', va='bottom', fontsize=10)

# ---------- Plot A: Profiles with SD split ----------
plot_profiles_for_svtype(df, "INS")
plot_profiles_for_svtype(df, "DEL")
```

```python
# ==== G4 stability in SV flanks: SD vs non-SD, binned violin plots ====

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

# ---------------- CONFIG ----------------
file_path   = "/home/alextu/scratch/segdup_scripts/all_combined_quadron_results_SV_flanks_withSDs.csv"
FLANK_LEN   = 2000     # bp (your window size)
BIN_SIZE    = 100      # bp per bin (try 50/100/200)
STABLE_Q    = 19       # threshold line
MIN_PER_BIN = 10       # drop bin×SD cells with < N obs from violin plots
SV_ORDER    = ["INS", "DEL"]
FLANK_ORDER = ["pre_flank", "post_flank"]
SD_LEVELS   = ["in_SD", "not_in_SD"]   # required SD label spellings
# ----------------------------------------

# ---------- Load & basic clean ----------
df = pd.read_csv(file_path)

# keep only rows with usable Q/L/POS
df = df.dropna(subset=["Q", "L"])
df["POS"] = pd.to_numeric(df["POS"], errors="coerce")
df["L"]   = pd.to_numeric(df["L"],   errors="coerce")
df = df.dropna(subset=["POS", "L"]).copy()

# midpoint within the flank (0..FLANK_LEN)
df["Midpoint"] = (df["POS"] + (df["POS"] + df["L"])) / 2

# ---------- Derive Flank & SVType ----------
def extract_flank(sample_name: str):
    m = re.search(r"_(pre|post)_flank(?:\.txt)?$", str(sample_name))
    return f"{m.group(1)}_flank" if m else "Unknown"

def extract_svtype(sample_name: str):
    m = re.search(r"(?:^|[_\-.])(INS|DEL)(?:[_\-.]|$)", str(sample_name), flags=re.IGNORECASE)
    return m.group(1).upper() if m else "Unknown"

df["Flank"]  = df["Sample"].apply(extract_flank)
df["SVType"] = df["Sample"].apply(extract_svtype)
df = df[df["Flank"].isin(FLANK_ORDER) & df["SVType"].isin(SV_ORDER)].copy()

# ---------- Normalize SD labels ----------
df["SD"] = df["SD"].map(lambda x: "in_SD" if str(x).lower()=="in_sd"
                                   else ("not_in_SD" if str(x).lower()=="not_in_sd" else "NA"))
df = df[df["SD"].isin(SD_LEVELS)].copy()

# ---------- Bin by position within the flank ----------
# clip to [0, FLANK_LEN) just in case
df["Midpoint"] = df["Midpoint"].clip(lower=0, upper=FLANK_LEN - 1e-6)

bin_edges  = np.arange(0, FLANK_LEN + BIN_SIZE, BIN_SIZE)
bin_labels = [f"{bin_edges[i]}-{bin_edges[i+1]}" for i in range(len(bin_edges)-1)]
df["Bin"]  = pd.cut(df["Midpoint"], bins=bin_edges, right=False, labels=bin_labels)
df["Bin"]  = pd.Categorical(df["Bin"], categories=bin_labels, ordered=True)

# optional: drop super-sparse (SVType×Flank×SD×Bin) cells
cell_counts = (df.groupby(["SVType","Flank","SD","Bin"]).size()
                 .reset_index(name="n"))
valid_cells = cell_counts[cell_counts["n"] >= MIN_PER_BIN][["SVType","Flank","SD","Bin"]]
df_v = df.merge(valid_cells, on=["SVType","Flank","SD","Bin"], how="inner")

# ---------- Plotting helper ----------
def violin_by_bins(df_in, svtype_label):
    sub = df_in[df_in["SVType"] == svtype_label].copy()
    if sub.empty:
        print(f"[WARN] No data for {svtype_label}; skipping.")
        return

    fig, axs = plt.subplots(1, 2, figsize=(18, 6), sharey=True)
    fig.suptitle(f"{svtype_label} — G4 Stability (Q) by Position Bin and SD Context",
                 fontsize=18, fontweight="bold")

    for i, flank in enumerate(FLANK_ORDER):
        ax = axs[i]
        pane = sub[sub["Flank"] == flank]
        if pane.empty:
            ax.set_visible(False)
            continue

        sns.violinplot(
            data=pane, x="Bin", y="Q", hue="SD",
            dodge=True, cut=0, inner="quartile", ax=ax
        )
        ax.axhline(y=STABLE_Q, color="black", linestyle="--", linewidth=1)
        ax.set_title(flank.replace("_"," ").title(), fontsize=16)
        ax.set_xlabel("Position bin within flank (bp)", fontsize=13)
        if i == 0:
            ax.set_ylabel("G4 Stability (Q)", fontsize=13)
        else:
            ax.set_ylabel("")
        ax.tick_params(axis="x", labelrotation=90)
        ax.grid(True, axis="y", linestyle="--", alpha=0.35)

        # keep one legend for the whole figure
        if i == 1:
            ax.legend(title="SD", fontsize=10, title_fontsize=11)
        else:
            ax.legend_.remove()

    plt.tight_layout()
    plt.show()

# ---------- MAKE THE PLOTS ----------
violin_by_bins(df_v, "INS")
violin_by_bins(df_v, "DEL")
```

```python
# ==== G4 stability in SV flanks: SD vs non-SD, 5-bin violins + per-bin stats (balanced) ====

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from scipy.stats import mannwhitneyu

# ---------------- CONFIG ----------------
file_path      = "/home/alextu/scratch/segdup_scripts/all_combined_quadron_results_SV_flanks_withSDs.csv"
FLANK_LEN      = 2000        # bp (your window size)
N_BINS         = 5           # force exactly 5 bins
STABLE_Q       = 19          # threshold line
MIN_PER_BIN    = 10          # for plotting: drop bin×SD cells with < N obs
MIN_PER_GROUP  = 5           # for stats: require at least this many per group per bin
SV_ORDER       = ["INS", "DEL"]
FLANK_ORDER    = ["pre_flank", "post_flank"]
SD_LEVELS      = ["in_SD", "not_in_SD"]
OUT_CSV        = "/home/alextu/scratch/segdup_scripts/g4_flanks_5bin_sd_stats_balanced.csv"
RNG_SEED       = 42          # reproducible downsampling
# ----------------------------------------

# ---------- Load & clean ----------
df = pd.read_csv(file_path)
df = df.dropna(subset=["Q", "L"])
df["POS"] = pd.to_numeric(df["POS"], errors="coerce")
df["L"]   = pd.to_numeric(df["L"],   errors="coerce")
df = df.dropna(subset=["POS", "L"]).copy()
df["Midpoint"] = (df["POS"] + (df["POS"] + df["L"])) / 2

def extract_flank(s):
    m = re.search(r"_(pre|post)_flank(?:\.txt)?$", str(s))
    return f"{m.group(1)}_flank" if m else "Unknown"

def extract_svtype(s):
    m = re.search(r"(?:^|[_\-.])(INS|DEL)(?:[_\-.]|$)", str(s), flags=re.IGNORECASE)
    return m.group(1).upper() if m else "Unknown"

df["Flank"]  = df["Sample"].apply(extract_flank)
df["SVType"] = df["Sample"].apply(extract_svtype)
df = df[df["Flank"].isin(FLANK_ORDER) & df["SVType"].isin(SV_ORDER)].copy()

def norm_sd(x):
    s = str(x).strip().lower()
    if s == "in_sd": return "in_SD"
    if s == "not_in_sd": return "not_in_SD"
    return "NA"

df["SD"] = df["SD"].map(norm_sd)
df = df[df["SD"].isin(SD_LEVELS)].copy()

# ---------- 5 bins over [0, FLANK_LEN) ----------
df["Midpoint"] = df["Midpoint"].clip(0, FLANK_LEN-1e-6)
bin_edges  = np.linspace(0, FLANK_LEN, N_BINS+1)
bin_labels = [f"{int(bin_edges[i])}-{int(bin_edges[i+1])}" for i in range(N_BINS)]
df["Bin"]  = pd.cut(df["Midpoint"], bins=bin_edges, right=False, labels=bin_labels)
df["Bin"]  = pd.Categorical(df["Bin"], categories=bin_labels, ordered=True)

# ---------- NEW: downsample not_in_SD to match in_SD per (SVType, Flank, Bin) ----------
def downsample_not_in_sd(cell_df: pd.DataFrame) -> pd.DataFrame:
    a = cell_df[cell_df["SD"] == "in_SD"]
    b = cell_df[cell_df["SD"] == "not_in_SD"]
    if a.empty or b.empty:
        # need both groups to compare; if only one exists, return as-is
        return cell_df

    k = len(a)
    if len(b) <= k:
        # already balanced or not_in_SD is smaller
        return pd.concat([a, b], ignore_index=True)

    b_s = b.sample(n=k, replace=False, random_state=RNG_SEED)
    return pd.concat([a, b_s], ignore_index=True)

df_bal = (
    df.groupby(["SVType", "Flank", "Bin"], observed=False, group_keys=False)
      .apply(downsample_not_in_sd)
      .reset_index(drop=True)
)

# ---------- Drop sparse cells for plotting ----------
cell_counts = (
    df_bal.groupby(["SVType","Flank","SD","Bin"], observed=False).size()
          .reset_index(name="n")
)
valid_cells = cell_counts[cell_counts["n"] >= MIN_PER_BIN][["SVType","Flank","SD","Bin"]]
df_v = df_bal.merge(valid_cells, on=["SVType","Flank","SD","Bin"], how="inner")

# ---------- Stats per bin (on balanced data) ----------
records = []
for sv in SV_ORDER:
    for flank in FLANK_ORDER:
        sub = df_bal[(df_bal["SVType"]==sv) & (df_bal["Flank"]==flank)]
        if sub.empty:
            continue
        for b in bin_labels:
            pane = sub[sub["Bin"]==b]
            a = pane[pane["SD"]=="in_SD"]["Q"].dropna()
            b_ = pane[pane["SD"]=="not_in_SD"]["Q"].dropna()
            n1, n2 = len(a), len(b_)
            if n1>=MIN_PER_GROUP and n2>=MIN_PER_GROUP:
                u, p = mannwhitneyu(a, b_, alternative="two-sided")
            else:
                u, p = np.nan, np.nan
            records.append({
                "SVType": sv, "Flank": flank, "Bin": b,
                "n_in_SD": n1, "n_not_in_SD": n2,
                "median_in_SD": float(a.median()) if n1 else np.nan,
                "median_not_in_SD": float(b_.median()) if n2 else np.nan,
                "U": u, "p_value": p
            })

stats_df = pd.DataFrame.from_records(records)

# ---------- Benjamini–Hochberg FDR (index-safe) ----------
def bh_fdr(pvals):
    p = np.array(pvals, dtype=float)
    mask = ~np.isnan(p)
    m = mask.sum()
    out = np.full_like(p, np.nan, dtype=float)
    if m == 0:
        return out
    ranks = np.argsort(np.where(mask, p, np.inf))
    cummin = np.inf
    k = 0
    for i, idx in enumerate(ranks, start=1):
        if not mask[idx]:
            continue
        k += 1
        val = p[idx] * m / k
        cummin = min(cummin, val)
        out[idx] = min(cummin, 1.0)
    return out

def bh_on_series(s: pd.Series) -> pd.Series:
    adj = bh_fdr(s.values)
    return pd.Series(adj, index=s.index)

if not stats_df.empty:
    stats_df["p_value_adj"] = (
        stats_df.groupby(["SVType","Flank"], observed=False, group_keys=False)["p_value"]
                .apply(bh_on_series)
    )

# Save stats
stats_df.to_csv(OUT_CSV, index=False)
print(f"[INFO] Wrote stats to: {OUT_CSV}")
print(stats_df.sort_values(["SVType","Flank","Bin"]))

# ---------- Plot with significance stars (balanced data) ----------
def violin_by_bins(df_in, svtype_label, stats_df):
    sub = df_in[df_in["SVType"] == svtype_label].copy()
    if sub.empty:
        print(f"[WARN] No data for {svtype_label}; skipping.")
        return
    
    stats_sub = stats_df[stats_df["SVType"]==svtype_label]
    if stats_sub.empty:
        print(f"[WARN] No stats for {svtype_label}; skipping stars.")
    
    fig, axs = plt.subplots(1, 2, figsize=(18, 6), sharey=True)
    fig.suptitle(f"{svtype_label} — G4 Stability by 5 Position Bins (Balanced)", fontsize=18, fontweight="bold")
    
    for i, flank in enumerate(FLANK_ORDER):
        ax = axs[i]
        pane = sub[sub["Flank"] == flank]
        if pane.empty:
            ax.set_visible(False)
            continue
        
        sns.violinplot(
            data=pane, x="Bin", y="Q", hue="SD",
            dodge=True, cut=0, inner="quartile", ax=ax
        )
        ax.axhline(y=STABLE_Q, color="black", linestyle="--", linewidth=1)
        ax.set_title(flank.replace("_"," ").title())
        ax.tick_params(axis="x", labelrotation=0)
        if i==0: ax.set_ylabel("G4 Stability (Q)")
        else:    ax.set_ylabel("")
        
        flank_stats = stats_sub[stats_sub["Flank"]==flank]
        y_max = pane["Q"].max() if not pane.empty else 0
        y_annot = y_max + 2
        for j, bin_label in enumerate(df["Bin"].cat.categories):
            stat_row = flank_stats[flank_stats["Bin"]==bin_label]
            if stat_row.empty:
                continue
            p_adj = stat_row["p_value_adj"].values[0]
            star = "*" if (p_adj < 0.05) else "ns"
            ax.text(j, y_annot, star, ha="center", va="bottom", fontsize=12, color="red")
        
        if i==1: 
            ax.legend(title="SD")
        else:
            leg = ax.get_legend()
            if leg is not None: leg.remove()
    
    plt.tight_layout()
    plt.show()

# ---------- MAKE THE PLOTS ----------
violin_by_bins(df_v, "INS", stats_df)
violin_by_bins(df_v, "DEL", stats_df)
```
