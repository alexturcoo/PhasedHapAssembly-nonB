# Extracting Haplotype Specific SV Flanks
This script extracts haplotype specific flanks (2000bp up and downstream of SV)

```python
### DID NOT END UP USING THIS CODE CUZ IT WAS TOO SLOW HOWEVER IT WAS WORKING
### TOO SLOW WITH PYTHON WRAPPERS TO EXTRACT SEQUENCE
### USING NEW CODE TO DO THIS WITH ONLY UNIX COMMANDS AND SAMTOOLS CONSENSUS

import pysam
import csv
import os

# Define the input BAM file and output CSV file
bam_file = '/home/alextu/scratch/filtered_chr_bams/chm13/20240201_verkko_batch3_align_chm13/HG00096/HG00096_hap1_primary.bam'
output_csv = "/home/alextu/projects/def-sushant/alextu/PhasedHapAssembly-nonB/sv_flanks.csv"

# Check if the BAM file exists
if not os.path.exists(bam_file):
    raise FileNotFoundError(f"BAM file not found: {bam_file}")

# Open the BAM file using pysam
try:
    bam = pysam.AlignmentFile(bam_file, "rb")
except Exception as e:
    raise RuntimeError(f"Failed to open BAM file: {e}")

# Define the region flanking the structural variant - ALL TESTS
#region = {"chr": "chrY", "pre_start": 999374, "pre_end": 1001374, "post_start": 1001374, "post_end": 1003374} # - chrY test Working

region = {"chr": "chrX", "pre_start": 10273, "pre_end": 12273, "post_start": 13027, "post_end": 15027}  # Test1 - WORKING DEL chrX
#region = {"chr": "chrX", "pre_start": 21195, "pre_end": 23195, "post_start": 23500, "post_end": 25500}  # Test2 - WORKING DEL chrX
#region = {"chr": "chrX", "pre_start": 22839, "pre_end": 24839, "post_start": 24900, "post_end": 26900}  # Test3 - WORKING DEL chrX
#region = {"chr": "chrX", "pre_start": 3776, "pre_end": 5776, "post_start": 5776, "post_end": 7776}     # Test4 - WORKING INS chrX
#region = {"chr": "chrX", "pre_start": 8724, "pre_end": 10724, "post_start": 10724, "post_end": 12724}  # Test5 - WORKING INS chrX
#region = {"chr": "chrX", "pre_start": 56667, "pre_end": 58667, "post_start": 58667, "post_end": 60667} # Test6 - WORKING INS chrX

#region = {"chr": "chr22", "pre_start": 10043976, "pre_end": 10045976 , "post_start": 10045976, "post_end": 10047976} #chr22 Test - WORKING

#region = {"chr": "chr8", "pre_start": 100268293, "pre_end": 100270293 , "post_start": 100270609, "post_end": 100272609} #chr8 Test - Working, took a little long
#region = {"chr": "chr10", "pre_start": 105582257, "pre_end": 105584257 , "post_start": 105584315, "post_end": 105586315}

#region = {"chr": "chr21", "pre_start": 1023201, "pre_end": 1025201 , "post_start": 1025201, "post_end": 1027201} #chr21 Test NOT WORKING, Pre and Post Flank Doubled
#region = {"chr": "chr2", "pre_start": 1000, "pre_end": 3000 , "post_start": 3000, "post_end": 5000}    # Test - TAKES LONG chr2
#region = {"chr": "chr1", "pre_start": 0, "pre_end": 1727 , "post_start": 2442, "post_end": 4442}       # Test - Region not in bam chr1
#region = {"chr": "chr10", "pre_start": 100011800, "pre_end": 100013800 , "post_start": 100013800, "post_end": 100015800} # Test6 - Takes long


# Target sequence length
target_length = 2000

# Function to fetch a sequence from BAM file within a specific range
def get_sequence(bam, chrom, start, end):
    try:
        sequence = []
        for pileupcolumn in bam.pileup(chrom, start, end, truncate=True):
            for pileupread in pileupcolumn.pileups:
                if not pileupread.is_del and not pileupread.is_refskip:
                    sequence.append(pileupread.alignment.query_sequence[pileupread.query_position])
        return ''.join(sequence)
    except ValueError:
        # Return an empty sequence if out of range or missing
        print(f"Region {chrom}:{start}-{end} is out of range or missing in the BAM file")
        return None

# Adjust the start position for the pre-flanking sequence if it is shorter than target length
def adjust_pre_flank(bam, chrom, pre_start, pre_end, target_length):
    pre_flank_seq = get_sequence(bam, chrom, pre_start, pre_end)
    if pre_flank_seq is None:
        return "Region not in BAM", pre_start, pre_end
    
    pre_flank_seq_length = len(pre_flank_seq)
    
    # While the sequence is shorter than the target, keep adjusting
    while pre_flank_seq_length < target_length:
        missing_bases = target_length - pre_flank_seq_length
        pre_start -= missing_bases
        pre_flank_seq = get_sequence(bam, chrom, pre_start, pre_end)
        if pre_flank_seq is None:
            return "Region not in BAM", pre_start, pre_end
        pre_flank_seq_length = len(pre_flank_seq)
        print(pre_flank_seq_length)
    
    return pre_flank_seq, pre_start, pre_end

# Adjust the end position for the post-flanking sequence if it is shorter than target length
def adjust_post_flank(bam, chrom, post_start, post_end, target_length):
    post_flank_seq = get_sequence(bam, chrom, post_start, post_end)
    if post_flank_seq is None:
        return "Region not in BAM", post_start, post_end

    post_flank_seq_length = len(post_flank_seq)
    
    # While the sequence is shorter than the target, keep adjusting
    while post_flank_seq_length < target_length:
        missing_bases = target_length - post_flank_seq_length
        post_end += missing_bases
        post_flank_seq = get_sequence(bam, chrom, post_start, post_end)
        if post_flank_seq is None:
            return "Region not in BAM", post_start, post_end
        post_flank_seq_length = len(post_flank_seq)
    
    return post_flank_seq, post_start, post_end

# Fetch pre-flanking sequence (2000bp before the variant)
pre_flank_seq, adjusted_pre_start, adjusted_pre_end = adjust_pre_flank(bam, region["chr"], region["pre_start"], region["pre_end"], target_length)

# Fetch post-flanking sequence (2000bp after the variant)
post_flank_seq, adjusted_post_start, adjusted_post_end = adjust_post_flank(bam, region["chr"], region["post_start"], region["post_end"], target_length)

# Prepare the CSV file
with open(output_csv, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    # Write the header
    csvwriter.writerow(['ID', 'pre_flank_start', 'pre_flank_end', 'pre_flank_seq', 'pre_flank_seq_length',
                        'post_flank_start', 'post_flank_end', 'post_flank_seq', 'post_flank_seq_length'])

    # Write the pre and post flank data to the CSV
    if pre_flank_seq != "Region not in BAM" and pre_flank_seq != "":
        pre_flank_seq_length = len(pre_flank_seq)
    else:
        pre_flank_seq_length = "N/A"

    if post_flank_seq != "Region not in BAM" and post_flank_seq != "":
        post_flank_seq_length = len(post_flank_seq)
    else:
        post_flank_seq_length = "N/A"
    
    csvwriter.writerow([region["chr"], adjusted_pre_start, adjusted_pre_end, pre_flank_seq, pre_flank_seq_length,
                        adjusted_post_start, adjusted_post_end, post_flank_seq, post_flank_seq_length])

# Output the results
print(f"Output written to {output_csv}")
print(f"Pre-flank sequence length: {pre_flank_seq_length}")
print(f"Pre-flank sequence (last 20bp): {pre_flank_seq[-20:]}")
print(f"Post-flank sequence length: {post_flank_seq_length}")
print(f"Post-flank sequence (first 20bp): {post_flank_seq[:20]}")
```
