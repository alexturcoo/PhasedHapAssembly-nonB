```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import sys

# ==========================
# 1. Load the CSV Data
# ==========================

# Define the path to your CSV file
CSV_FILE_PATH = '/home/alextu/scratch/extract_sv_flanks_2000bp/all_combined_quadron_results_SV_flanks.csv'

try:
    df = pd.read_csv(CSV_FILE_PATH)
    print(f"Successfully loaded data from '{CSV_FILE_PATH}'.")
except FileNotFoundError:
    print(f"Error: The file '{CSV_FILE_PATH}' was not found.")
    sys.exit(1)
except pd.errors.EmptyDataError:
    print(f"Error: The file '{CSV_FILE_PATH}' is empty.")
    sys.exit(1)
except pd.errors.ParserError:
    print(f"Error: The file '{CSV_FILE_PATH}' does not appear to be in CSV format or is malformed.")
    sys.exit(1)

# ==========================
# 2. Define Processing Function
# ==========================

def process_sv(df, sv_type):
    """
    Process the DataFrame for a specific SV type (INS or DEL).

    Parameters:
        df (pd.DataFrame): The original DataFrame containing all SVs.
        sv_type (str): The SV type to process ('INS' or 'DEL').

    Returns:
        tuple: (df_hap1, df_hap2) DataFrames for haplotype 1 and haplotype 2.
    """
    print(f"\nProcessing {sv_type} SVs...")
    
    # Define a pattern to identify the specific SV type
    sv_pattern = f'_{sv_type}'
    
    # Filter the DataFrame to include only rows corresponding to the specific SV type
    df_sv = df[df['Sample'].str.contains(sv_pattern, regex=True)].copy()
    
    # Check if any rows were found
    if df_sv.empty:
        print(f"No {sv_type} SVs found in the dataset. Skipping {sv_type} processing.")
        return None, None
    
    print(f"Number of rows for {sv_type} SVs: {len(df_sv)}")
    
    # Optionally, list unique SV identifiers
    unique_sv_identifiers = df_sv['Sample'].unique()
    print(f"Unique {sv_type} SV identifiers found: {len(unique_sv_identifiers)}")
    
    # ==========================
    # 3. Extract Haplotype and Flank Type
    # ==========================
    
    def extract_hap_flank(sample):
        """
        Extract haplotype and flank type from the Sample filename.

        Parameters:
            sample (str): The filename string from the 'Sample' column.

        Returns:
            tuple: (haplotype, flank_type) e.g., ('hap1', 'pre')
        """
        # Example Sample filename:
        # flanking_sequences_HG00733_hap2_chr10_1006437-1008457_chr10_1008457_59_INS_pre_flank.txt

        # Use regex to extract haplotype and flank type
        pattern = r'hap(\d+)_.*_(pre|post)_flank\.txt$'
        match = re.search(pattern, sample)
        if match:
            hap = 'hap' + match.group(1)
            flank = match.group(2)
            return hap, flank
        else:
            return None, None

    # Apply the extraction function to the 'Sample' column
    df_sv[['hap', 'flank']] = df_sv['Sample'].apply(
        lambda x: pd.Series(extract_hap_flank(x))
    )

    # Drop rows where haplotype or flank type couldn't be extracted
    initial_count = len(df_sv)
    df_sv.dropna(subset=['hap', 'flank'], inplace=True)
    dropped_count = initial_count - len(df_sv)
    if dropped_count > 0:
        print(f"Dropped {dropped_count} rows due to inability to extract haplotype or flank type.")

    print(f"After extracting haplotype and flank type, {len(df_sv)} rows remain.")

    # ==========================
    # 4. Calculate Relative Positions
    # ==========================

    # Define a function to calculate rel_pos based on flank type
    def calculate_rel_pos(row):
        """
        Calculate the relative position of the G4 structure.

        Parameters:
            row (pd.Series): A row of the DataFrame.

        Returns:
            int: Relative position in base pairs.
        """
        if row['flank'] == 'pre':
            return row['POS'] - 2000
        elif row['flank'] == 'post':
            return row['POS']
        else:
            return None  # For unexpected flank types

    # Apply the function to calculate 'rel_pos'
    df_sv['rel_pos'] = df_sv.apply(calculate_rel_pos, axis=1)

    # Drop rows where 'rel_pos' couldn't be calculated
    initial_count = len(df_sv)
    df_sv.dropna(subset=['rel_pos'], inplace=True)
    dropped_count = initial_count - len(df_sv)
    if dropped_count > 0:
        print(f"Dropped {dropped_count} rows due to inability to calculate relative position.")

    # ==========================
    # 5. Validate Relative Positions
    # ==========================

    # Define a function to validate 'rel_pos' based on 'flank' type
    def validate_rel_pos(row):
        if row['flank'] == 'pre':
            return -2000 <= row['rel_pos'] <= 0
        elif row['flank'] == 'post':
            return 0 <= row['rel_pos'] <= 2000
        else:
            return False  # For unexpected flank types

    # Apply validation
    invalid_rel_pos = df_sv[~df_sv.apply(validate_rel_pos, axis=1)]
    if not invalid_rel_pos.empty:
        print(f"Warning: {len(invalid_rel_pos)} rows have inconsistent 'rel_pos' values based on 'flank' type.")
        # Optionally, inspect these rows
        print(invalid_rel_pos[['Sample', 'rel_pos', 'flank']].head())
        # Drop the invalid rows
        df_sv = df_sv[df_sv.apply(validate_rel_pos, axis=1)]

    print(f"After validating 'rel_pos' values, {len(df_sv)} rows remain.")

    # ==========================
    # 6. Separate Haplotype 1 and Haplotype 2
    # ==========================

    # Filter data for hap1 and hap2
    df_hap1 = df_sv[df_sv['hap'] == 'hap1']
    df_hap2 = df_sv[df_sv['hap'] == 'hap2']

    print(f"Number of G4 structures for hap1: {len(df_hap1)}")
    print(f"Number of G4 structures for hap2: {len(df_hap2)}")

    # Optionally, report the number of SVs included
    num_sv_hap1 = df_hap1['Sample'].nunique()
    num_sv_hap2 = df_hap2['Sample'].nunique()
    print(f"Number of unique {sv_type} SVs for hap1: {num_sv_hap1}")
    print(f"Number of unique {sv_type} SVs for hap2: {num_sv_hap2}")

    return df_hap1, df_hap2

# ==========================
# 3. Process Insertion (INS) SVs
# ==========================

df_hap1_ins, df_hap2_ins = process_sv(df, 'INS')

# ==========================
# 4. Process Deletion (DEL) SVs
# ==========================

df_hap1_del, df_hap2_del = process_sv(df, 'DEL')

# ==========================
# 5. Visualize the Density of G4 Structures for INS and DEL
# ==========================

# Set the visual style
sns.set(style="whitegrid")

# Initialize the matplotlib figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(28, 8), sharey=True)

# --------------------------
# Plot for Insertion (INS) SVs
# --------------------------
if df_hap1_ins is not None and df_hap2_ins is not None:
    sns.kdeplot(
        data=df_hap1_ins,
        x='rel_pos',
        label='hap1',
        fill=True,
        common_norm=False,
        alpha=0.5,
        color='blue',
        ax=axes[0]
    )
    
    sns.kdeplot(
        data=df_hap2_ins,
        x='rel_pos',
        label='hap2',
        fill=True,
        common_norm=False,
        alpha=0.5,
        color='orange',
        ax=axes[0]
    )
    
    axes[0].axvline(0, color='black', linestyle='--', linewidth=1)
    axes[0].set_title("INS SVs: Density of G-Quadruplex Structures Relative to SV Breakpoints", fontsize=16)
    axes[0].set_xlabel('Position Relative to SV Breakpoint (bp)', fontsize=14)
    axes[0].set_ylabel('Density', fontsize=14)
    axes[0].set_xlim(-2000, 2000)
    axes[0].legend(title='Haplotype', fontsize=12, title_fontsize=12)
    axes[0].grid(True, linestyle='--', alpha=0.5)
else:
    axes[0].text(0.5, 0.5, 'No INS SV data available', horizontalalignment='center', verticalalignment='center', fontsize=14)
    axes[0].set_title("INS SVs: No Data Available", fontsize=16)
    axes[0].set_xlabel('Position Relative to SV Breakpoint (bp)', fontsize=14)
    axes[0].set_ylabel('Density', fontsize=14)
    axes[0].set_xlim(-2000, 2000)
    axes[0].grid(True, linestyle='--', alpha=0.5)

# --------------------------
# Plot for Deletion (DEL) SVs
# --------------------------
if df_hap1_del is not None and df_hap2_del is not None:
    sns.kdeplot(
        data=df_hap1_del,
        x='rel_pos',
        label='hap1',
        fill=True,
        common_norm=False,
        alpha=0.5,
        color='green',
        ax=axes[1]
    )
    
    sns.kdeplot(
        data=df_hap2_del,
        x='rel_pos',
        label='hap2',
        fill=True,
        common_norm=False,
        alpha=0.5,
        color='red',
        ax=axes[1]
    )
    
    axes[1].axvline(0, color='black', linestyle='--', linewidth=1)
    axes[1].set_title("DEL SVs: Density of G-Quadruplex Structures Relative to SV Breakpoints", fontsize=16)
    axes[1].set_xlabel('Position Relative to SV Breakpoint (bp)', fontsize=14)
    axes[1].set_xlim(-2000, 2000)
    axes[1].legend(title='Haplotype', fontsize=12, title_fontsize=12)
    axes[1].grid(True, linestyle='--', alpha=0.5)
else:
    axes[1].text(0.5, 0.5, 'No DEL SV data available', horizontalalignment='center', verticalalignment='center', fontsize=14)
    axes[1].set_title("DEL SVs: No Data Available", fontsize=16)
    axes[1].set_xlabel('Position Relative to SV Breakpoint (bp)', fontsize=14)
    axes[1].set_xlim(-2000, 2000)
    axes[1].grid(True, linestyle='--', alpha=0.5)

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# ==========================
# 1. Load the CSV Data
# ==========================
CSV_FILE_PATH = '/home/alextu/projects/def-sushant/alextu/extract_SV_and_MEI_flank_scripts/all_combined_quadron_results_MEI_flanks_with_TE_designations.csv'

try:
    df = pd.read_csv(CSV_FILE_PATH)
    print(f"Successfully loaded data from '{CSV_FILE_PATH}'.")
except Exception as e:
    print(f"Error: {e}")
    exit()

# ==========================
# 2. Extract Flank Type
# ==========================
def extract_flank(sample):
    """Extract flank type (pre or post) from the Sample filename."""
    pattern = r'(pre|post)_flank'
    match = re.search(pattern, sample)
    return match.group(1) if match else None

df['flank'] = df['Sample'].apply(extract_flank)
df.dropna(subset=['flank'], inplace=True)

# ==========================
# 3. Calculate Relative Position
# ==========================
def calculate_rel_pos(row):
    """Calculate relative position:
    - Pre-flank: Maps START closer to 2000 as closer to 0 (breakpoint), into -2000 to 0.
    - Post-flank: Position naturally increases from 0 to 2000.
    """
    if row['flank'] == 'pre':
        return  row['POS'] - 2000  # Shift pre-flank into -2000 to 0
    elif row['flank'] == 'post':
        return row['POS']         # Post-flank stays as is (0 to 2000)
    return None

df['rel_pos'] = df.apply(calculate_rel_pos, axis=1)
df.dropna(subset=['rel_pos'], inplace=True)

# ==========================
# 4. Plot KDE Density by TE Designation
# ==========================
# Set seaborn style
sns.set(style="whitegrid")

# Get unique TE designations
te_designations = df['te_designation'].unique()

# Initialize subplots
num_te = len(te_designations)
fig, axes = plt.subplots(1, num_te, figsize=(10 * num_te, 6), sharey=True)

# Ensure axes is a list for single TE designation
if num_te == 1:
    axes = [axes]

# Plot each TE designation
for ax, te in zip(axes, te_designations):
    te_df = df[df['te_designation'] == te]
    
    # Filter for upstream and downstream
    pre_df = te_df[te_df['flank'] == 'pre']
    post_df = te_df[te_df['flank'] == 'post']
    
    # Plot KDE for upstream (pre-flank) clipped to -2000 to 0
    if not pre_df['rel_pos'].empty and pre_df['rel_pos'].nunique() > 1:
        sns.kdeplot(
            data=pre_df, 
            x='rel_pos', 
            label='Upstream', 
            fill=False, 
            ax=ax, 
            color='blue', 
            alpha=0.5, 
            clip=(-2000, 0)  # Ensures KDE is bounded to -2000 to 0
        )
    
    # Plot KDE for downstream (post-flank) clipped to 0 to 2000
    if not post_df['rel_pos'].empty and post_df['rel_pos'].nunique() > 1:
        sns.kdeplot(
            data=post_df, 
            x='rel_pos', 
            label='Downstream', 
            fill=False, 
            ax=ax, 
            color='orange', 
            alpha=0.5, 
            clip=(0, 2000)  # Ensures KDE is bounded to 0 to 2000
        )
    
    # Add a vertical line at 0
    ax.axvline(0, color='black', linestyle='--', linewidth=1)
    
    # Titles and labels
    ax.set_title(f"{te} G-quadruplex Density", fontsize=14)
    ax.set_xlabel("Position Relative to Breakpoint (bp)", fontsize=12)
    ax.set_ylabel("Density", fontsize=12)
    ax.legend(title="Flank Type", fontsize=10, title_fontsize=10)
    ax.grid(True, linestyle='--', alpha=0.5)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# ==========================
# 1. Load the CSV Data
# ==========================
CSV_FILE_PATH = '/home/alextu/projects/def-sushant/alextu/extract_SV_and_MEI_flank_scripts/MEIflanks_collapsed_summary_metrics_chr_position_with_TE_designations.csv'

try:
    df = pd.read_csv(CSV_FILE_PATH)
    print(f"Successfully loaded data from '{CSV_FILE_PATH}'.")
except Exception as e:
    print(f"Error: {e}")
    exit()

# ==========================
# 2. Extract Haplotype and Flank Type
# ==========================
def extract_flank(sample):
    """Extract flank type (pre or post) from the Sample filename."""
    pattern = r'(pre|post)_flank'
    match = re.search(pattern, sample)
    if match:
        return match.group(1)
    return None

df['flank'] = df['Sample_Haplotype_Chromosome'].apply(lambda x: extract_flank(x))
df.dropna(subset=['flank'], inplace=True)

# ==========================
# 3. Calculate Relative Position
# ==========================
def calculate_rel_pos(row):
    """
    Calculate relative position such that:
    - Pre-flank (upstream): Position closer to 2000 maps to 0 (breakpoint).
    - Post-flank (downstream): Position naturally increases from 0.
    """
    if row['flank'] == 'pre':
        return row['Start'] - 2000  # Flip the axis for upstream
    elif row['flank'] == 'post':
        return row['Start']  # Downstream stays as is
    return None

# Apply calculate_rel_pos to create 'rel_pos' column
df['rel_pos'] = df.apply(calculate_rel_pos, axis=1)
df.dropna(subset=['rel_pos'], inplace=True)

# Verify rel_pos column
print(df[['flank', 'Start', 'rel_pos']].head())

# ==========================
# 4. Plot Density for Each Motif Type and TE Designation
# ==========================
# Set seaborn style
sns.set(style="whitegrid")

# Get unique TE designations and Motif Types
te_designations = df['te_designation'].unique()
motif_types = df['Motif Type'].unique()

# Initialize subplots
num_motifs = len(motif_types)
fig, axes = plt.subplots(len(te_designations), num_motifs, figsize=(5 * num_motifs, 4 * len(te_designations)), sharey=False)

# Ensure axes is a 2D array
if len(te_designations) == 1:
    axes = axes.reshape(1, -1)
if num_motifs == 1:
    axes = axes.reshape(-1, 1)

# Plot densities for each TE designation and motif type
for row_idx, te in enumerate(te_designations):
    te_df = df[df['te_designation'] == te]
    for col_idx, motif in enumerate(motif_types):
        motif_df = te_df[te_df['Motif Type'] == motif]
        
        # Filter for upstream and downstream
        pre_df = motif_df[motif_df['flank'] == 'pre']
        post_df = motif_df[motif_df['flank'] == 'post']
        
        # Select appropriate axes
        ax = axes[row_idx, col_idx]
        
        # Track if any data was plotted
        plotted = False
        
        # Plot upstream density (clip to -2000 to 0)
        if not pre_df['rel_pos'].empty and pre_df['rel_pos'].nunique() > 1:
            sns.kdeplot(
                data=pre_df, 
                x='rel_pos', 
                label='Upstream', 
                fill=True, 
                ax=ax, 
                color='blue', 
                alpha=0.5, 
                clip=(-2000, 0)
            )
            plotted = True
        
        # Plot downstream density (clip to 0 to 2000)
        if not post_df['rel_pos'].empty and post_df['rel_pos'].nunique() > 1:
            sns.kdeplot(
                data=post_df, 
                x='rel_pos', 
                label='Downstream', 
                fill=True, 
                ax=ax, 
                color='orange', 
                alpha=0.5, 
                clip=(0, 2000)
            )
            plotted = True
        
        # Add legend only if data was plotted
        if plotted:
            ax.legend(title="Flank Type", fontsize=8, title_fontsize=8)
        
        # Plot vertical line at 0
        ax.axvline(0, color='black', linestyle='--', linewidth=1)
        ax.set_title(f"{motif} Density in {te}", fontsize=10)
        ax.set_xlabel("Position Relative to Breakpoint (bp)", fontsize=8)
        ax.set_ylabel("Density", fontsize=8)
        ax.grid(True, linestyle='--', alpha=0.5)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import sys

# ==========================
# Configuration
# ==========================
DEBUG = True  # Set to False to disable debug print statements

# File paths
MEI_FILE_PATH = '/home/alextu/projects/def-sushant/alextu/extract_SV_and_MEI_flank_scripts/MEIflanks_collapsed_summary_metrics_chr_position_with_TE_designations.csv'
G4_FILE_PATH = '/home/alextu/projects/def-sushant/alextu/extract_SV_and_MEI_flank_scripts/all_combined_quadron_results_MEI_flanks_with_TE_designations.csv'

# Plot configurations
PLOT_SIZE = (10, 6)
X_LIMITS = (-1800, 1800)  # Adjusted to ensure symmetry
PLOT_SAVE_DIR = '/home/alextu/projects/def-sushant/'  # Update as needed

# ==========================
# Utility Functions
# ==========================
def debug_print(*args):
    """Utility function for conditional debugging."""
    if DEBUG:
        print(*args)

# ==========================
# Load CSV Data
# ==========================
def load_csv(file_path, description):
    """Load a CSV file and handle errors."""
    debug_print(f"Loading {description} from '{file_path}'...")
    try:
        df = pd.read_csv(file_path)
        debug_print(f"Successfully loaded {description}. Shape: {df.shape}")
        debug_print(f"{description} Data Sample:")
        debug_print(df.head())
        return df
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
        sys.exit(1)
    except pd.errors.EmptyDataError:
        print(f"Error: The file '{file_path}' is empty.")
        sys.exit(1)
    except pd.errors.ParserError:
        print(f"Error: The file '{file_path}' does not appear to be in CSV format or is malformed.")
        sys.exit(1)

df_mei = load_csv(MEI_FILE_PATH, "MEI data")
df_g4 = load_csv(G4_FILE_PATH, "G4 data")

# ==========================
# Adjust rel_pos for Centering
# ==========================
def adjust_rel_pos(row, col_name, desired_pre_min, desired_pre_max, desired_post_min, desired_post_max):
    """
    Adjust the relative positions to be within the range -1800 to 0 for pre-flank
    and 0 to 1800 for post-flank using linear scaling.
    """
    if row['flank'] == 'pre':
        # Scale pre-flank range
        return row[col_name] - 2000
    elif row['flank'] == 'post':
        # Scale post-flank range
        return row[col_name] - 1
    return None

# ==========================
# Process MEI Data
# ==========================
def process_mei_data(df):
    """
    Process the MEI DataFrame to extract TE designations, motif types, flanks, and calculate relative positions.
    """
    debug_print("\nProcessing MEI data...")

    # Extract flank type (pre/post)
    df['flank'] = df['Sample_Haplotype_Chromosome'].str.extract(r'(pre|post)_flank')

    # Drop rows without flank information
    df.dropna(subset=['flank'], inplace=True)

    # Adjust relative positions
    df['rel_pos'] = df.apply(
        adjust_rel_pos,
        axis=1,
        col_name='Start',
        desired_pre_min=-1800,
        desired_pre_max=0,
        desired_post_min=0,
        desired_post_max=1800
    )

    # Validate ranges
    debug_print(f"Adjusted MEI Pre-flank rel_pos range: {df[df['flank'] == 'pre']['rel_pos'].min()} to {df[df['flank'] == 'pre']['rel_pos'].max()}")
    debug_print(f"Adjusted MEI Post-flank rel_pos range: {df[df['flank'] == 'post']['rel_pos'].min()} to {df[df['flank'] == 'post']['rel_pos'].max()}")

    df = df[(df['rel_pos'] >= X_LIMITS[0]) & (df['rel_pos'] <= X_LIMITS[1])]  # Filter by limits

    debug_print(f"Processed MEI Data Shape: {df.shape}")
    debug_print("Processed MEI Data Sample:")
    debug_print(df.head())

    return df

df_mei = process_mei_data(df_mei)

# ==========================
# Process G4 Data
# ==========================
def process_g4_data(df):
    """
    Process the G4 DataFrame to extract flank type and calculate relative positions.
    """
    debug_print("\nProcessing G4 data...")

    # Extract flank type (pre/post)
    df['flank'] = df['Sample'].str.extract(r'(pre|post)_flank')

    # Drop rows without flank information
    df.dropna(subset=['flank'], inplace=True)

    # Adjust relative positions
    df['rel_pos'] = df.apply(
        adjust_rel_pos,
        axis=1,
        col_name='POS',
        desired_pre_min=-1800,
        desired_pre_max=0,
        desired_post_min=0,
        desired_post_max=1800
    )

    # Validate ranges
    debug_print(f"Adjusted G4 Pre-flank rel_pos range: {df[df['flank'] == 'pre']['rel_pos'].min()} to {df[df['flank'] == 'pre']['rel_pos'].max()}")
    debug_print(f"Adjusted G4 Post-flank rel_pos range: {df[df['flank'] == 'post']['rel_pos'].min()} to {df[df['flank'] == 'post']['rel_pos'].max()}")

    df = df[(df['rel_pos'] >= X_LIMITS[0]) & (df['rel_pos'] <= X_LIMITS[1])]  # Filter by limits

    debug_print(f"Processed G4 Data Shape: {df.shape}")
    debug_print("Processed G4 Data Sample:")
    debug_print(df.head())

    return df

df_g4 = process_g4_data(df_g4)

# ==========================
# Generate Density Plots
# ==========================
def plot_density_per_te(df_mei, df_g4, te_designation, save_dir=PLOT_SAVE_DIR):
    """
    Plot density for a single TE designation, including G4 data.
    """
    debug_print(f"\nGenerating plot for TE Designation: {te_designation}")

    # Filter data
    df_te = df_mei[df_mei['te_designation'] == te_designation]
    df_te_g4 = df_g4[df_g4['te_designation'] == te_designation]

    debug_print(f"MEI Data filtered: {df_te.shape[0]} rows")
    debug_print(f"G4 Data filtered: {df_te_g4.shape[0]} rows")

    unique_motifs = df_te['Motif Type'].unique()
    debug_print(f"Unique Motif Types: {unique_motifs}")

    # Assign colors to motifs
    palette = sns.color_palette(n_colors=len(unique_motifs) + 1)
    motif_colors = dict(zip(unique_motifs, palette))
    motif_colors['G4'] = palette[-1]

    plt.figure(figsize=PLOT_SIZE)

    # Plot motif densities
    if not df_te.empty:
        for motif in unique_motifs:
            motif_df = df_te[df_te['Motif Type'] == motif]
            if motif_df.empty:
                continue
            sns.kdeplot(
                data=motif_df,
                x='rel_pos',
                label=motif,
                linewidth=1.5,
                color=motif_colors[motif],
                bw_adjust=0.6  # Smoothing
            )

    # Plot G4 densities
    if not df_te_g4.empty:
        sns.kdeplot(
            data=df_te_g4,
            x='rel_pos',
            label='G4',
            linewidth=1.5,
            linestyle='-',
            color='green',
            bw_adjust=0.6
        )

    # Add plot details
    plt.axvline(0, color='black', linestyle='--', linewidth=1)  # Reference line at 0
    plt.title(f"TE Designation: {te_designation} - Density of Motifs and G4", fontsize=16)
    plt.xlabel('Position Relative to Breakpoint (bp)', fontsize=14)
    plt.ylabel('Density', fontsize=14)
    plt.xlim(X_LIMITS)
    plt.legend(title='Motif Type', fontsize=10, title_fontsize=12)
    plt.tight_layout()
    plt.show()

# Define the TE designations to plot
te_designations = df_mei['te_designation'].unique()

# Generate and save plots for each TE designation
for te in te_designations:
    plot_density_per_te(df_mei, df_g4, te)

debug_print("\nDensity plots successfully created and saved.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
import sys

# ==========================
# Configuration
# ==========================
DEBUG = True  # Set to False to disable debug print statements

# File paths
MEI_FILE_PATH = '/home/alextu/projects/def-sushant/alextu/extract_SV_and_MEI_flank_scripts/MEIflanks_collapsed_summary_metrics_chr_position_with_TE_designations.csv'
G4_FILE_PATH = '/home/alextu/projects/def-sushant/alextu/extract_SV_and_MEI_flank_scripts/all_combined_quadron_results_MEI_flanks_with_TE_designations.csv'

# Plot configurations
PLOT_SIZE = (10, 6)
X_LIMITS = (-2000, 2000)
BIN_SIZE = 100  # Size of each bin in bp
PLOT_SAVE_DIR = '/home/alextu/projects/def-sushant/'  # Update as needed

# ==========================
# Utility Functions
# ==========================
def debug_print(*args):
    """Utility function for conditional debugging."""
    if DEBUG:
        print(*args)

# ==========================
# 1. Load the CSV Data
# ==========================
def load_csv(file_path, description):
    """Load a CSV file and handle errors."""
    debug_print(f"Loading {description} from '{file_path}'...")
    try:
        df = pd.read_csv(file_path)
        debug_print(f"Successfully loaded {description}. Shape: {df.shape}")
        debug_print(f"{description} Data Sample:")
        debug_print(df.head())
        return df
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
        sys.exit(1)
    except pd.errors.EmptyDataError:
        print(f"Error: The file '{file_path}' is empty.")
        sys.exit(1)
    except pd.errors.ParserError:
        print(f"Error: The file '{file_path}' does not appear to be in CSV format or is malformed.")
        sys.exit(1)

df_mei = load_csv(MEI_FILE_PATH, "MEI data")
df_g4 = load_csv(G4_FILE_PATH, "G4 data")

# ==========================
# 2. Process MEI Data
# ==========================
def process_mei_data(df):
    """
    Process the MEI DataFrame to extract TE designations, motif types, flanks, and calculate relative positions.
    """
    debug_print("\nProcessing MEI data...")

    # Extract flank type (pre/post)
    df['flank'] = df['Sample_Haplotype_Chromosome'].str.extract(r'(pre|post)_flank')

    # Drop rows without flank information
    df.dropna(subset=['flank'], inplace=True)

    # Calculate relative position
    def calculate_rel_pos(row, col_name='Start'):
        if row['flank'] == 'pre':
            return row[col_name] - 2001  # Align pre-flanks: 1..1989 -> -1989..-1
        elif row['flank'] == 'post':
            return row[col_name]  # Post-flanks remain as is
        return None

    df['rel_pos'] = df.apply(calculate_rel_pos, axis=1)

    # Bin the data
    df['binned_pos'] = pd.cut(df['rel_pos'], bins=np.arange(X_LIMITS[0], X_LIMITS[1] + BIN_SIZE, BIN_SIZE), labels=False)
    df = df[(df['rel_pos'] >= X_LIMITS[0]) & (df['rel_pos'] <= X_LIMITS[1])]  # Filter by limits

    debug_print(f"Processed MEI Data Shape: {df.shape}")
    debug_print("Processed MEI Data Sample:")
    debug_print(df.head())

    return df

df_mei = process_mei_data(df_mei)

# ==========================
# 3. Process G4 Data
# ==========================
def process_g4_data(df):
    """
    Process the G4 DataFrame to extract flank type and calculate relative positions.
    """
    debug_print("\nProcessing G4 data...")

    # Extract flank type (pre/post)
    df['flank'] = df['Sample'].str.extract(r'(pre|post)_flank')

    # Drop rows without flank information
    df.dropna(subset=['flank'], inplace=True)

    # Calculate relative position
    def calculate_rel_pos(row, col_name='POS'):
        if row['flank'] == 'pre':
            return row[col_name] - 2001  # Align pre-flanks: 1..1989 -> -1989..-1
        elif row['flank'] == 'post':
            return row[col_name]  # Post-flanks remain as is
        return None

    df['rel_pos'] = df.apply(calculate_rel_pos, axis=1)

    # Bin the data
    df['binned_pos'] = pd.cut(df['rel_pos'], bins=np.arange(X_LIMITS[0], X_LIMITS[1] + BIN_SIZE, BIN_SIZE), labels=False)
    df = df[(df['rel_pos'] >= X_LIMITS[0]) & (df['rel_pos'] <= X_LIMITS[1])]  # Filter by limits

    debug_print(f"Processed G4 Data Shape: {df.shape}")
    debug_print("Processed G4 Data Sample:")
    debug_print(df.head())

    return df

df_g4 = process_g4_data(df_g4)

# ==========================
# 4. Generate Density Plots
# ==========================
def plot_density_per_te(df_mei, df_g4, te_designation, save_dir=PLOT_SAVE_DIR):
    """
    Plot density for a single TE designation, including G4 data.
    """
    debug_print(f"\nGenerating plot for TE Designation: {te_designation}")

    # Filter data
    df_te = df_mei[df_mei['te_designation'] == te_designation]
    df_te_g4 = df_g4[df_g4['te_designation'] == te_designation]

    debug_print(f"MEI Data filtered: {df_te.shape[0]} rows")
    debug_print(f"G4 Data filtered: {df_te_g4.shape[0]} rows")

    unique_motifs = df_te['Motif Type'].unique()
    debug_print(f"Unique Motif Types: {unique_motifs}")

    # Assign colors to motifs
    palette = sns.color_palette(n_colors=len(unique_motifs) + 1)
    motif_colors = dict(zip(unique_motifs, palette))
    motif_colors['G4'] = palette[-1]

    plt.figure(figsize=PLOT_SIZE)

    # Plot motif densities
    if not df_te.empty:
        for motif in unique_motifs:
            motif_df = df_te[df_te['Motif Type'] == motif]
            binned_data = motif_df.groupby('binned_pos')['rel_pos'].count()  # Aggregate by bins
            sns.lineplot(x=binned_data.index * BIN_SIZE + X_LIMITS[0], y=binned_data.values, label=motif, linewidth=1.5)

    # Plot G4 densities
    if not df_te_g4.empty:
        binned_g4 = df_te_g4.groupby('binned_pos')['rel_pos'].count()
        sns.lineplot(x=binned_g4.index * BIN_SIZE + X_LIMITS[0], y=binned_g4.values, label='G4', linewidth=1.5, linestyle='--', color='green')

    # Add plot details
    plt.axvline(0, color='black', linestyle='--', linewidth=1)  # Reference line at 0
    plt.title(f"TE Designation: {te_designation} - Density of Motifs and G4", fontsize=16)
    plt.xlabel('Position Relative to Breakpoint (bp)', fontsize=14)
    plt.ylabel('Density', fontsize=14)
    plt.xlim(X_LIMITS)
    plt.legend(title='Motif Type', fontsize=10, title_fontsize=12)
    plt.tight_layout()
    plt.show()

# Define the TE designations to plot
te_designations = df_mei['te_designation'].unique()

# Generate and save plots for each TE designation
for te in te_designations:
    plot_density_per_te(df_mei, df_g4, te)

debug_print("\nDensity plots successfully created and saved.")
```
