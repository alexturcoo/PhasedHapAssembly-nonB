# Analysis of SV and MEI Flanks

I extracted 2000bp flanks up and downstream haplotype specific SVs and MEIs - we are trying to understand enrichment patterns of non-Bs in flanks of these SVs and MEIs

# SVs

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flank data
flank_data = pd.read_csv('/home/alextu/projects/def-sushant/verkko_batch123_nonB_annotations_aligned_july2024/SVflanks_collapsed_summary_metrics_chr.csv')

# Extract the pre and post flank information
flank_data['Flank_Type'] = flank_data['Sample_Haplotype_Chromosome'].apply(lambda x: 'pre' if 'pre_flank' in x else 'post')

# Extract the SV type (INS or DEL) from the Sample_Haplotype_Chromosome column
flank_data['SV_Type'] = flank_data['Sample_Haplotype_Chromosome'].apply(lambda x: 'INS' if 'INS' in x else 'DEL' if 'DEL' in x else 'Other')

# Load the new G-quadruplex data
g_quad_data = pd.read_csv('/home/alextu/scratch/aggregated_sv_coverage.csv')

# Extract pre/post and INS/DEL from the `Sample` column
g_quad_data['Flank_Type'] = g_quad_data['Sample'].apply(lambda x: 'pre' if 'pre_flank' in x else 'post')
g_quad_data['SV_Type'] = g_quad_data['Sample'].apply(lambda x: 'INS' if 'INS' in x else 'DEL' if 'DEL' in x else 'Other')

# Add G-quadruplex as the motif type
g_quad_data['Motif Type'] = 'G-Quadruplex'

# Calculate overall G-quadruplex metrics
g_quad_summary = g_quad_data.groupby('Motif Type').agg(
    Avg_Bases_Covered=('Total_Length', 'mean'),
    Max_Bases_Covered=('Total_Length', 'max'),
    Total_Flanks=('Total_Length', 'count')
).reset_index()

# Combine flank and G-quad data for overall summary
overall_summary = flank_data.groupby('Motif Type').agg(
    Avg_Bases_Covered=('Total Bases Covered', 'mean'),
    Max_Bases_Covered=('Total Bases Covered', 'max'),
    Total_Flanks=('Flank_Type', 'count')
).reset_index()
overall_summary_combined = pd.concat([overall_summary, g_quad_summary], ignore_index=True)

# Sort overall summary by Avg_Bases_Covered
overall_summary_combined = overall_summary_combined.sort_values(by='Avg_Bases_Covered', ascending=False)

# Plot 1: Overall Summary
plt.figure(figsize=(8, 8))
ax = sns.barplot(
    x='Motif Type',
    y='Avg_Bases_Covered',
    data=overall_summary_combined,
    order=overall_summary_combined['Motif Type'].unique(),
    errorbar=None
)
plt.title('Average Bases Covered by Motif Type (No Stratification)')
plt.xlabel('Motif Type')
plt.ylabel('Average Bases Covered')
plt.xticks(rotation=45)
plt.tight_layout()

# Annotate averages on bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10)

plt.show()

# Stratify by Flank_Type and SV_Type
flank_summary = flank_data.groupby(['Flank_Type', 'Motif Type', 'SV_Type']).agg(
    Avg_Bases_Covered=('Total Bases Covered', 'mean'),
    Max_Bases_Covered=('Total Bases Covered', 'max'),
    Total_Flanks=('Flank_Type', 'count')
).reset_index()

# Aggregate G-quadruplex data by Flank_Type and SV_Type
g_quad_flank_summary = g_quad_data.groupby(['Flank_Type', 'Motif Type', 'SV_Type']).agg(
    Avg_Bases_Covered=('Total_Length', 'mean'),
    Max_Bases_Covered=('Total_Length', 'max'),
    Total_Flanks=('Total_Length', 'count')
).reset_index()

flank_summary_combined = pd.concat([flank_summary, g_quad_flank_summary], ignore_index=True)

# Sort flank_summary_combined by Avg_Bases_Covered
flank_summary_combined = flank_summary_combined.sort_values(by='Avg_Bases_Covered', ascending=False)

# Plot 2: Stratified by Flank_Type
plt.figure(figsize=(8, 8))
ax = sns.barplot(
    x='Motif Type',
    y='Avg_Bases_Covered',
    hue='Flank_Type',
    data=flank_summary_combined,
    order=flank_summary_combined['Motif Type'].unique(),
    errorbar=None
)
plt.title('Average Bases Covered by Motif Type (Pre vs Post Flanks)')
plt.xlabel('Motif Type')
plt.ylabel('Average Bases Covered')
plt.xticks(rotation=45)
plt.tight_layout()

# Annotate averages on bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10)

plt.show()

# Plot 3: Stratified by SV_Type
plt.figure(figsize=(8, 8))
ax = sns.barplot(
    x='Motif Type',
    y='Avg_Bases_Covered',
    hue='SV_Type',
    data=flank_summary_combined,
    order=flank_summary_combined['Motif Type'].unique(),
    errorbar=None
)
plt.title('Average Bases Covered by Motif Type (INS vs DEL)')
plt.xlabel('Motif Type')
plt.ylabel('Average Bases Covered')
plt.xticks(rotation=45)
plt.tight_layout()

# Annotate averages on bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10)

plt.show()

# Combine Flank_Type and SV_Type for detailed metrics
flank_data['Flank_SV_Type'] = flank_data['Flank_Type'] + '_' + flank_data['SV_Type']
g_quad_data['Flank_SV_Type'] = g_quad_data['Flank_Type'] + '_' + g_quad_data['SV_Type']

flank_combined_summary = flank_data.groupby(['Flank_SV_Type', 'Motif Type']).agg(
    Avg_Bases_Covered=('Total Bases Covered', 'mean'),
    Max_Bases_Covered=('Total Bases Covered', 'max'),
    Total_Flanks=('Flank_SV_Type', 'count')
).reset_index()

g_quad_combined_summary = g_quad_data.groupby(['Flank_SV_Type', 'Motif Type']).agg(
    Avg_Bases_Covered=('Total_Length', 'mean'),
    Max_Bases_Covered=('Total_Length', 'max'),
    Total_Flanks=('Total_Length', 'count')
).reset_index()

combined_summary_combined = pd.concat([flank_combined_summary, g_quad_combined_summary], ignore_index=True)

# Sort combined_summary_combined by Avg_Bases_Covered
combined_summary_combined = combined_summary_combined.sort_values(by='Avg_Bases_Covered', ascending=False)

# Plot 4: Stratified by Flank_SV_Type
plt.figure(figsize=(8, 8))
ax = sns.barplot(
    x='Motif Type',
    y='Avg_Bases_Covered',
    hue='Flank_SV_Type',
    data=combined_summary_combined,
    order=combined_summary_combined['Motif Type'].unique(),
    errorbar=None
)
plt.title('Average Bases Covered by Motif Type (Pre_INS, Post_INS, Pre_DEL, Post_DEL)')
plt.xlabel('Motif Type')
plt.ylabel('Average Bases Covered')
plt.xticks(rotation=45)
plt.tight_layout()

# Annotate averages on bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Load the data
data = pd.read_csv('/home/alextu/projects/def-sushant/verkko_batch123_nonB_annotations_aligned_july2024/SVflanks_collapsed_summary_metrics_chr.csv')

# Extract the haplotype part from Sample_Haplotype_Chromosome
data['Haplotype'] = data['Sample_Haplotype_Chromosome'].apply(
    lambda x: re.search(r'(HG\d{5}_hap\d|NA\d{5}_hap\d)', x).group(0) if re.search(r'(HG\d{5}_hap\d|NA\d{5}_hap\d)', x) else None
)

# Identify the SV type (INS or DEL)
data['SV_Type'] = data['Sample_Haplotype_Chromosome'].apply(lambda x: 'INS' if 'INS' in x else 'DEL' if 'DEL' in x else 'Other')

# Extract chromosome information from Sample_Haplotype_Chromosome
data['Chromosome'] = data['Sample_Haplotype_Chromosome'].apply(lambda x: re.search(r'chr[\dXY]+', x).group(0) if re.search(r'chr[\dXY]+', x) else None)

# Identify fully covered flanks (2000 bases)
fully_covered_flanks = data[data['Total Bases Covered'] >= 1999]

# Count the number of fully covered flanks by motif type
fully_covered_counts = fully_covered_flanks['Motif Type'].value_counts().reset_index()
fully_covered_counts.columns = ['Motif Type', 'Count']

# Print unique haplotypes for each motif type
haplotypes_per_motif = fully_covered_flanks.groupby('Motif Type')['Haplotype'].unique().reset_index()
haplotypes_per_motif.columns = ['Motif Type', 'Unique Haplotypes']

# Merge the count data with the haplotypes
fully_covered_summary = pd.merge(fully_covered_counts, haplotypes_per_motif, on='Motif Type')

# Plot the number of fully covered flanks by motif type
plt.figure(figsize=(8, 8))
sns.barplot(x='Motif Type', y='Count', data=fully_covered_summary, errorbar=None)
plt.title('Number of Fully Covered Flanks (2000 Bases) by Motif Type in Structural Variant Flanks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print out summary of fully covered flanks with unique haplotypes for review
print("Summary of flanks completely covered by non-B DNA motifs (2000 bases):")
print(fully_covered_summary[['Motif Type', 'Count', 'Unique Haplotypes']])

# Additional: Count the number of fully covered flanks by motif type and SV type
sv_counts = fully_covered_flanks.groupby(['Motif Type', 'SV_Type']).size().reset_index(name='Count')

# Plot the number of fully covered flanks by motif type, stratified by SV_Type
plt.figure(figsize=(8, 8))
sns.barplot(x='Motif Type', y='Count', hue='SV_Type', data=sv_counts, errorbar=None)
plt.title('Number of Fully Covered Flanks (2000 Bases) by Motif Type, Stratified by SV Type (INS vs DEL)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Additional: Count the number of fully covered flanks by chromosome and motif type
chromosome_motif_counts = fully_covered_flanks.groupby(['Chromosome', 'Motif Type']).size().unstack(fill_value=0)

# Plot the stacked bar plot
chromosome_motif_counts.plot(kind='bar', stacked=True, figsize=(8, 8), colormap='viridis')
plt.title('Number of Fully Covered Flanks (2000 Bases) by Chromosome, Stratified by Motif Type')
plt.xlabel('Chromosome')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Load the data
data = pd.read_csv('/home/alextu/projects/def-sushant/verkko_batch123_nonB_annotations_aligned_july2024/SVflanks_collapsed_summary_metrics_chr.csv')

# Extract the haplotype part from Sample_Haplotype_Chromosome
data['Haplotype'] = data['Sample_Haplotype_Chromosome'].apply(
    lambda x: re.search(r'(HG\d{5}_hap\d|NA\d{5}_hap\d)', x).group(0) if re.search(r'(HG\d{5}_hap\d|NA\d{5}_hap\d)', x) else None
)

# Extract the correct sample name (HG##### or NA####)
data['Sample'] = data['Sample_Haplotype_Chromosome'].str.extract(r'(HG\d+|NA\d+)')[0]

# Load metadata
metadata_path = '/home/alextu/scratch/igsr_sample_metadata/igsr_samples.tsv'  # Update this with the actual path
metadata_df = pd.read_csv(metadata_path, delimiter='\t')

# Merge metadata with the main data to add Superpopulation name
data = pd.merge(data, metadata_df[['Sample name', 'Superpopulation name']], left_on='Sample', right_on='Sample name', how='left')

# Identify fully covered flanks (2000 bases)
fully_covered_flanks = data[data['Total Bases Covered'] >= 1999]

# Count the number of unique samples in each superpopulation
sample_counts = data[['Sample', 'Superpopulation name']].drop_duplicates().groupby('Superpopulation name').size().reset_index(name='Sample Count')

# Count the number of fully covered flanks by motif type and superpopulation
fully_covered_counts = fully_covered_flanks.groupby(['Motif Type', 'Superpopulation name']).size().reset_index(name='Count')

# Merge the sample counts with the fully covered counts
fully_covered_counts = pd.merge(fully_covered_counts, sample_counts, on='Superpopulation name')

# Normalize the counts by the number of samples in each superpopulation
fully_covered_counts['Normalized Count'] = fully_covered_counts['Count'] / fully_covered_counts['Sample Count']

# Print unique haplotypes for each motif type and superpopulation
haplotypes_per_motif = fully_covered_flanks.groupby(['Motif Type', 'Superpopulation name'])['Haplotype'].unique().reset_index()
haplotypes_per_motif.columns = ['Motif Type', 'Superpopulation name', 'Unique Haplotypes']

# Merge the normalized count data with the haplotypes
fully_covered_summary = pd.merge(fully_covered_counts, haplotypes_per_motif, on=['Motif Type', 'Superpopulation name'])

# Plot the normalized number of fully covered flanks by motif type and superpopulation
plt.figure(figsize=(16, 8))
sns.barplot(x='Motif Type', y='Normalized Count', hue='Superpopulation name', data=fully_covered_summary, errorbar=None)
plt.title('Normalized Number of Fully Covered Flanks (2000 Bases) by Motif Type Across Superpopulations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print out summary of fully covered flanks with unique haplotypes for review
print("Normalized summary of flanks completely covered by non-B DNA motifs (2000 bases) with superpopulation:")
print(fully_covered_summary[['Motif Type', 'Superpopulation name', 'Normalized Count', 'Unique Haplotypes']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the main data
data = pd.read_csv('/home/alextu/projects/def-sushant/verkko_batch123_nonB_annotations_aligned_july2024/MEIflanks_collapsed_summary_metrics_chr_with_TE_designations.csv')

# Extract the correct sample name (HG##### or NA####)
data['Sample'] = data['Sample_Haplotype_Chromosome'].str.extract(r'(HG\d+|NA\d+)')[0]

# Extract the pre and post flank information
data['Flank_Type'] = data['Sample_Haplotype_Chromosome'].apply(lambda x: 'pre' if 'pre_flank' in x else 'post')

# Load the G-quadruplex data
gquad_data = pd.read_csv('/home/alextu/scratch/aggregated_meiflank_coverage_gquads.csv')

# Extract the correct sample name for G-quadruplex data
gquad_data['Sample'] = gquad_data['Sample'].str.extract(r'(HG\d+|NA\d+)')[0]
gquad_data['Motif Type'] = 'G-quadruplex'  # Add G-quadruplex as a motif type

# Ensure G-quadruplex data has the necessary columns
gquad_data = gquad_data.rename(columns={'Total_Length': 'Total Bases Covered'})

# Add the G-quadruplex data into the main dataframe
data = pd.concat([data, gquad_data[['Sample', 'te_designation', 'Motif Type', 'Total Bases Covered']]], ignore_index=True)

# Filter out rows with Motif Type as snRNA
data = data[data['te_designation'] != 'snRNA']

# Overall metrics with no stratification by Flank_Type or TE Designation
overall_summary_no_strat = data.groupby('Motif Type').agg(
    Avg_Bases_Covered=('Total Bases Covered', 'mean')
).reset_index()

# Sort by Avg_Bases_Covered for plotting
overall_summary_no_strat_sorted = overall_summary_no_strat.sort_values(by='Avg_Bases_Covered', ascending=False)

# Plot the average bases covered across the motif types using Matplotlib
fig, ax = plt.subplots(figsize=(8, 8))
bars = ax.bar(
    overall_summary_no_strat_sorted['Motif Type'],
    overall_summary_no_strat_sorted['Avg_Bases_Covered'],
    color='skyblue'
)

# Annotate bars with the average base coverage values
for bar, value in zip(bars, overall_summary_no_strat_sorted['Avg_Bases_Covered']):
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # Center the text
        bar.get_height() + 0.05,  # Slightly above the bar
        f"{value:.2f}",  # Format the value
        ha='center',  # Horizontal alignment
        va='bottom',  # Vertical alignment
        fontsize=10
    )

# Add plot title and labels
ax.set_title('Average Bases Covered by Motif Type', fontsize=16)
ax.set_xlabel('Motif Type', fontsize=14)
ax.set_ylabel('Average Bases Covered', fontsize=14)
ax.set_xticks(range(len(overall_summary_no_strat_sorted['Motif Type'])))
ax.set_xticklabels(overall_summary_no_strat_sorted['Motif Type'], rotation=45, ha='right', fontsize=12)

# Adjust layout
plt.tight_layout()
plt.show()

# Group data by Motif Type and te_designation
te_summary = data.groupby(['Motif Type', 'te_designation']).agg(
    Avg_Bases_Covered=('Total Bases Covered', 'mean')
).reset_index()

# Add overall average for sorting and sort the data
overall_avg = te_summary.groupby('Motif Type')['Avg_Bases_Covered'].mean().reset_index()
sorted_motifs = overall_avg.sort_values(by='Avg_Bases_Covered', ascending=False)['Motif Type']

# Pivot the data for grouped bar chart
te_pivot = te_summary.pivot(index='Motif Type', columns='te_designation', values='Avg_Bases_Covered').fillna(0)
te_pivot = te_pivot.loc[sorted_motifs]  # Reorder rows by sorted motif order

# Create a grouped bar chart
fig, ax = plt.subplots(figsize=(10, 8))
x = np.arange(len(te_pivot.index))  # X locations for groups
width = 0.2  # Width of bars

# Plot each te_designation as a separate bar in the group
for i, col in enumerate(te_pivot.columns):
    ax.bar(
        x + i * width,
        te_pivot[col],
        width,
        label=col
    )

# Add annotations
for i, col in enumerate(te_pivot.columns):
    for j, value in enumerate(te_pivot[col]):
        ax.text(
            x[j] + i * width,
            value + 0.05,
            f"{value:.2f}",
            ha='center',
            va='bottom',
            fontsize=10
        )

# Add plot details
ax.set_title('Average Bases Covered by Motif Type and TE Designation', fontsize=16)
ax.set_xlabel('Motif Type', fontsize=14)
ax.set_ylabel('Average Bases Covered', fontsize=14)
ax.set_xticks(x + width * (len(te_pivot.columns) - 1) / 2)
ax.set_xticklabels(te_pivot.index, rotation=45, ha='right', fontsize=12)
ax.legend(title='TE Designation')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('/home/alextu/projects/def-sushant/verkko_batch123_nonB_annotations_aligned_july2024/MEIflanks_collapsed_summary_metrics_chr_with_TE_designations.csv')

# Filter for fully covered flanks (2000 bases)
fully_covered_flanks = data[data['Total Bases Covered'] >= 500]

# Count the number of fully covered flanks by motif type and TE designation
fully_covered_counts = fully_covered_flanks.groupby(['Motif Type', 'te_designation']).size().reset_index(name='Count')

# Print unique haplotypes for each motif type and TE designation
haplotypes_per_motif = fully_covered_flanks.groupby(['Motif Type', 'te_designation'])['Sample_Haplotype_Chromosome'].unique().reset_index()
haplotypes_per_motif.columns = ['Motif Type', 'te_designation', 'Unique Haplotypes']

# Merge the count data with the haplotypes
fully_covered_summary = pd.merge(fully_covered_counts, haplotypes_per_motif, on=['Motif Type', 'te_designation'])

# Sort by TE designation for better visualization
fully_covered_summary_sorted = fully_covered_summary.sort_values(by='te_designation')

# Plot the number of fully covered flanks by motif type and TE designation
plt.figure(figsize=(8, 8))
sns.barplot(
    x='Motif Type',
    y='Count',
    hue='te_designation',
    data=fully_covered_summary_sorted,
    errorbar=None
)
plt.title('Number of Fully Covered MEI Flanks (500 Bases) by Motif Type and TE Designation')
plt.xlabel('Motif Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print out the summary of fully covered flanks with unique haplotypes for review
print("Summary of fully covered MEI flanks (1000 bases) sorted by TE designation:")
print(fully_covered_summary_sorted[['Motif Type', 'te_designation', 'Count', 'Unique Haplotypes']])
```
