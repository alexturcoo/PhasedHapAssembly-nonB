# Analysis of SV and MEI Flanks

I extracted 2000bp flanks up and downstream haplotype specific SVs and MEIs - we are trying to understand enrichment patterns of non-Bs in flanks of these SVs and MEIs

# SVs

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/home/alextu/projects/def-sushant/verkko_batch123_nonB_annotations_aligned_july2024/SVflanks_collapsed_summary_metrics_chr.csv')

# Extract the pre and post flank information
data['Flank_Type'] = data['Sample_Haplotype_Chromosome'].apply(lambda x: 'pre' if 'pre_flank' in x else 'post')

# Extract the SV type (INS or DEL) from the Sample_Haplotype_Chromosome column
data['SV_Type'] = data['Sample_Haplotype_Chromosome'].apply(lambda x: 'INS' if 'INS' in x else 'DEL' if 'DEL' in x else 'Other')

# Basic metrics stratified by Motif Type (no stratification by Flank_Type or SV_Type)
overall_summary = data.groupby('Motif Type').agg(
    Total_Flanks=('Flank_Type', 'count'),
    Avg_Total_Count=('Total Count', 'mean'),
    Avg_Bases_Covered=('Total Bases Covered', 'mean'),
    Max_Bases_Covered=('Total Bases Covered', 'max')
).reset_index()

# Sort the overall summary by Avg_Bases_Covered in descending order for plotting
overall_summary_sorted = overall_summary.sort_values(by='Avg_Bases_Covered', ascending=False)

# Plot the average bases covered across the motifs (no stratification)
plt.figure(figsize=(8, 8))
sns.barplot(
    x='Motif Type',
    y='Avg_Bases_Covered',
    data=overall_summary_sorted,
    order=overall_summary_sorted['Motif Type'].unique(),  # Ensure consistent order
    errorbar=None
)
plt.title('Average Bases Covered by Motif Type (No Stratification) in Structral Variant Flanks')
plt.xlabel('Motif Type')
plt.ylabel('Average Bases Covered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Continue with your existing plots...

# Basic metrics stratified by Flank_Type, Motif Type, and SV_Type
flank_summary = data.groupby(['Flank_Type', 'Motif Type', 'SV_Type']).agg(
    Total_Flanks=('Flank_Type', 'count'),
    Avg_Total_Count=('Total Count', 'mean'),
    Avg_Bases_Covered=('Total Bases Covered', 'mean'),
    Max_Bases_Covered=('Total Bases Covered', 'max')
).reset_index()

# Sort the data by Avg_Bases_Covered in descending order for each plot
flank_summary_sorted = flank_summary.sort_values(by='Avg_Bases_Covered', ascending=False)

# Print the summary
print(flank_summary_sorted)

# Plot the average bases covered across the motifs for Flank_Type
plt.figure(figsize=(8, 8))
sns.barplot(
    x='Motif Type',
    y='Avg_Bases_Covered',
    hue='Flank_Type',
    data=flank_summary_sorted,
    order=flank_summary_sorted['Motif Type'].unique(),  # Ensure consistent order
    errorbar=None
)
plt.title('Average Bases Covered by Motif Type (Pre vs Post Flanks) in Structural Variant Flanks')
plt.xlabel('Motif Type')
plt.ylabel('Average Bases Covered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot the average bases covered across the motifs for SV_Type
plt.figure(figsize=(8, 8))
sns.barplot(
    x='Motif Type',
    y='Avg_Bases_Covered',
    hue='SV_Type',
    data=flank_summary_sorted,
    order=flank_summary_sorted['Motif Type'].unique(),  # Ensure consistent order
    errorbar=None
)
plt.title('Average Bases Covered by Motif Type (INS vs DEL) in Structural Variant Flanks')
plt.xlabel('Motif Type')
plt.ylabel('Average Bases Covered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Create a combined column for Flank_Type and SV_Type
data['Flank_SV_Type'] = data['Flank_Type'] + '_' + data['SV_Type']

# Basic metrics stratified by Flank_SV_Type and Motif Type
combined_summary = data.groupby(['Flank_SV_Type', 'Motif Type']).agg(
    Total_Flanks=('Flank_SV_Type', 'count'),
    Avg_Total_Count=('Total Count', 'mean'),
    Avg_Bases_Covered=('Total Bases Covered', 'mean'),
    Max_Bases_Covered=('Total Bases Covered', 'max')
).reset_index()

# Sort the combined summary by Avg_Bases_Covered in descending order
combined_summary_sorted = combined_summary.sort_values(by='Avg_Bases_Covered', ascending=False)

# Plot the average bases covered across the motifs for Flank_SV_Type
plt.figure(figsize=(8, 8))
sns.barplot(
    x='Motif Type',
    y='Avg_Bases_Covered',
    hue='Flank_SV_Type',
    data=combined_summary_sorted,
    order=combined_summary_sorted['Motif Type'].unique(),  # Ensure consistent order
    errorbar=None
)
plt.title('Average Bases Covered by Motif Type (Pre_INS, Post_INS, Pre_DEL, Post_DEL) in Structural Variant Flanks')
plt.xlabel('Motif Type')
plt.ylabel('Average Bases Covered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/home/alextu/projects/def-sushant/verkko_batch123_nonB_annotations_aligned_july2024/SVflanks_collapsed_summary_metrics_chr.csv')

# Extract the correct sample name (HG##### or NA####)
data['Sample'] = data['Sample_Haplotype_Chromosome'].str.extract(r'(HG\d+|NA\d+)')[0]

# Load metadata
metadata_path = '/home/alextu/scratch/igsr_sample_metadata/igsr_samples.tsv'  # Update this with the actual path
metadata_df = pd.read_csv(metadata_path, delimiter='\t')

# Merge metadata with the main data
merged_data = pd.merge(data, metadata_df[['Sample name', 'Superpopulation name']], left_on='Sample', right_on='Sample name', how='left')

# Categorize into "AFR" and "non-AFR"
merged_data['Superpopulation Group'] = merged_data['Superpopulation name'].apply(lambda x: 'AFR' if x == 'African Ancestry' else 'non-AFR')

# Basic metrics stratified only by Motif Type and Superpopulation Group
merged_summary = merged_data.groupby(['Motif Type', 'Superpopulation Group']).agg(
    Avg_Bases_Covered=('Total Bases Covered', 'mean')
).reset_index()

# Sort by Avg_Bases_Covered for plotting
merged_summary_sorted = merged_summary.sort_values(by='Avg_Bases_Covered', ascending=False)

# Print the merged summary
print(merged_summary_sorted)

# Plot the average bases covered by Motif Type across AFR and non-AFR superpopulations
plt.figure(figsize=(8, 8))
sns.barplot(
    x='Motif Type',
    y='Avg_Bases_Covered',
    hue='Superpopulation Group',
    data=merged_summary_sorted,
    errorbar=None
)
plt.title('Average Bases Covered by Motif Type Across AFR and non-AFR Superpopulations in SV flanks')
plt.xlabel('Motif Type')
plt.ylabel('Average Bases Covered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Boxplot to show the distribution of bases covered for each Motif Type by Superpopulation Group
plt.figure(figsize=(8, 8))
sns.boxplot(
    x='Motif Type',
    y='Total Bases Covered',
    hue='Superpopulation Group',
    data=merged_data,
    palette='Set2',
    showfliers=False
)
plt.title('Distribution of Total Bases Covered by Motif Type Across AFR and non-AFR Superpopulations in SV flanks (No Outliers)')
plt.xlabel('Motif Type')
plt.ylabel('Total Bases Covered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Boxplot to show the distribution of bases covered for each Motif Type by Superpopulation Group
plt.figure(figsize=(8, 8))
sns.boxplot(
    x='Motif Type',
    y='Total Bases Covered',
    hue='Superpopulation Group',
    data=merged_data,
    palette='Set2'
)
plt.title('Distribution of Total Bases Covered by Motif Type Across AFR and non-AFR Superpopulations in SV flanks (With Outliers)')
plt.xlabel('Motif Type')
plt.ylabel('Total Bases Covered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Load the data
data = pd.read_csv('/home/alextu/projects/def-sushant/verkko_batch123_nonB_annotations_aligned_july2024/SVflanks_collapsed_summary_metrics_chr.csv')

# Extract the haplotype part from Sample_Haplotype_Chromosome
data['Haplotype'] = data['Sample_Haplotype_Chromosome'].apply(
    lambda x: re.search(r'(HG\d{5}_hap\d|NA\d{5}_hap\d)', x).group(0) if re.search(r'(HG\d{5}_hap\d|NA\d{5}_hap\d)', x) else None
)

# Identify the SV type (INS or DEL)
data['SV_Type'] = data['Sample_Haplotype_Chromosome'].apply(lambda x: 'INS' if 'INS' in x else 'DEL' if 'DEL' in x else 'Other')

# Extract chromosome information from Sample_Haplotype_Chromosome
data['Chromosome'] = data['Sample_Haplotype_Chromosome'].apply(lambda x: re.search(r'chr[\dXY]+', x).group(0) if re.search(r'chr[\dXY]+', x) else None)

# Identify fully covered flanks (2000 bases)
fully_covered_flanks = data[data['Total Bases Covered'] >= 1999]

# Count the number of fully covered flanks by motif type
fully_covered_counts = fully_covered_flanks['Motif Type'].value_counts().reset_index()
fully_covered_counts.columns = ['Motif Type', 'Count']

# Print unique haplotypes for each motif type
haplotypes_per_motif = fully_covered_flanks.groupby('Motif Type')['Haplotype'].unique().reset_index()
haplotypes_per_motif.columns = ['Motif Type', 'Unique Haplotypes']

# Merge the count data with the haplotypes
fully_covered_summary = pd.merge(fully_covered_counts, haplotypes_per_motif, on='Motif Type')

# Plot the number of fully covered flanks by motif type
plt.figure(figsize=(8, 8))
sns.barplot(x='Motif Type', y='Count', data=fully_covered_summary, errorbar=None)
plt.title('Number of Fully Covered Flanks (2000 Bases) by Motif Type in Structural Variant Flanks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print out summary of fully covered flanks with unique haplotypes for review
print("Summary of flanks completely covered by non-B DNA motifs (2000 bases):")
print(fully_covered_summary[['Motif Type', 'Count', 'Unique Haplotypes']])

# Additional: Count the number of fully covered flanks by motif type and SV type
sv_counts = fully_covered_flanks.groupby(['Motif Type', 'SV_Type']).size().reset_index(name='Count')

# Plot the number of fully covered flanks by motif type, stratified by SV_Type
plt.figure(figsize=(8, 8))
sns.barplot(x='Motif Type', y='Count', hue='SV_Type', data=sv_counts, errorbar=None)
plt.title('Number of Fully Covered Flanks (2000 Bases) by Motif Type, Stratified by SV Type (INS vs DEL)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Additional: Count the number of fully covered flanks by chromosome and motif type
chromosome_motif_counts = fully_covered_flanks.groupby(['Chromosome', 'Motif Type']).size().unstack(fill_value=0)

# Plot the stacked bar plot
chromosome_motif_counts.plot(kind='bar', stacked=True, figsize=(8, 8), colormap='viridis')
plt.title('Number of Fully Covered Flanks (2000 Bases) by Chromosome, Stratified by Motif Type')
plt.xlabel('Chromosome')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Load the data
data = pd.read_csv('/home/alextu/projects/def-sushant/verkko_batch123_nonB_annotations_aligned_july2024/SVflanks_collapsed_summary_metrics_chr.csv')

# Extract the haplotype part from Sample_Haplotype_Chromosome
data['Haplotype'] = data['Sample_Haplotype_Chromosome'].apply(
    lambda x: re.search(r'(HG\d{5}_hap\d|NA\d{5}_hap\d)', x).group(0) if re.search(r'(HG\d{5}_hap\d|NA\d{5}_hap\d)', x) else None
)

# Extract the correct sample name (HG##### or NA####)
data['Sample'] = data['Sample_Haplotype_Chromosome'].str.extract(r'(HG\d+|NA\d+)')[0]

# Load metadata
metadata_path = '/home/alextu/scratch/igsr_sample_metadata/igsr_samples.tsv'  # Update this with the actual path
metadata_df = pd.read_csv(metadata_path, delimiter='\t')

# Merge metadata with the main data to add Superpopulation name
data = pd.merge(data, metadata_df[['Sample name', 'Superpopulation name']], left_on='Sample', right_on='Sample name', how='left')

# Identify fully covered flanks (2000 bases)
fully_covered_flanks = data[data['Total Bases Covered'] >= 1999]

# Count the number of unique samples in each superpopulation
sample_counts = data[['Sample', 'Superpopulation name']].drop_duplicates().groupby('Superpopulation name').size().reset_index(name='Sample Count')

# Count the number of fully covered flanks by motif type and superpopulation
fully_covered_counts = fully_covered_flanks.groupby(['Motif Type', 'Superpopulation name']).size().reset_index(name='Count')

# Merge the sample counts with the fully covered counts
fully_covered_counts = pd.merge(fully_covered_counts, sample_counts, on='Superpopulation name')

# Normalize the counts by the number of samples in each superpopulation
fully_covered_counts['Normalized Count'] = fully_covered_counts['Count'] / fully_covered_counts['Sample Count']

# Print unique haplotypes for each motif type and superpopulation
haplotypes_per_motif = fully_covered_flanks.groupby(['Motif Type', 'Superpopulation name'])['Haplotype'].unique().reset_index()
haplotypes_per_motif.columns = ['Motif Type', 'Superpopulation name', 'Unique Haplotypes']

# Merge the normalized count data with the haplotypes
fully_covered_summary = pd.merge(fully_covered_counts, haplotypes_per_motif, on=['Motif Type', 'Superpopulation name'])

# Plot the normalized number of fully covered flanks by motif type and superpopulation
plt.figure(figsize=(16, 8))
sns.barplot(x='Motif Type', y='Normalized Count', hue='Superpopulation name', data=fully_covered_summary, errorbar=None)
plt.title('Normalized Number of Fully Covered Flanks (2000 Bases) by Motif Type Across Superpopulations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Print out summary of fully covered flanks with unique haplotypes for review
print("Normalized summary of flanks completely covered by non-B DNA motifs (2000 bases) with superpopulation:")
print(fully_covered_summary[['Motif Type', 'Superpopulation name', 'Normalized Count', 'Unique Haplotypes']])
```
