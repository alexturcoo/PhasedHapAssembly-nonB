```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D

# === 1. Load and clean data
file_path = "/home/alextu/scratch/free_energy_IRS_verkko123_wholegenome_chm13/HG00096_hap2_IR_free_energy.tsv"
df = pd.read_csv(file_path, sep="\t").drop(columns=["Sequence"])

df['Repeat'] = pd.to_numeric(df['Repeat'], errors='coerce')
df['Spacer'] = pd.to_numeric(df['Spacer'], errors='coerce')
df['Free_Energy'] = pd.to_numeric(df['Free_Energy'], errors='coerce')
df = df.dropna(subset=['Repeat', 'Spacer', 'Free_Energy'])

# === 2. Fit OLS regression
X = sm.add_constant(df[['Repeat', 'Spacer']])
y = df['Free_Energy']
model = sm.OLS(y, X).fit()

# Print summary
print(model.summary())

# === 3. Standardized coefficients
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['Repeat', 'Spacer']])
lr = LinearRegression().fit(X_scaled, y)
print("\nStandardized Coefficients:")
for feature, coef in zip(['Repeat', 'Spacer'], lr.coef_):
    print(f"{feature}: {coef:.3f}")

# === 4. Residuals plot
sns.set(style="whitegrid")
sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True, color="blue")
plt.axhline(0, color='gray', linestyle='--')
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot (OLS Model)")
plt.tight_layout()
plt.show()

# === 5. 3D Regression Surface + Scatter
# Create grid for surface
repeat_range = np.linspace(df['Repeat'].min(), df['Repeat'].max(), 50)
spacer_range = np.linspace(df['Spacer'].min(), df['Spacer'].max(), 50)
R, S = np.meshgrid(repeat_range, spacer_range)

# Predict Free Energy over grid
grid_df = pd.DataFrame({'Repeat': R.ravel(), 'Spacer': S.ravel()})
grid_df_const = sm.add_constant(grid_df)
Z = model.predict(grid_df_const).values.reshape(R.shape)

# Clip observed values for better color scaling
capped_energy = np.clip(df['Free_Energy'], -100, 0)

# === Plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Surface from OLS model
surf = ax.plot_surface(R, S, Z, cmap='viridis', alpha=0.6, edgecolor='none')

# Observed points
sc = ax.scatter(df['Repeat'], df['Spacer'], df['Free_Energy'],
                c=capped_energy, cmap='coolwarm', s=5, alpha=0.6)

# Labels
ax.set_xlabel("Repeat Length")
ax.set_ylabel("Spacer Length")
ax.set_zlabel("Free Energy (kcal/mol)")
ax.set_title("OLS Regression Surface with Observed Data Points")

# Colorbar
fig.colorbar(sc, ax=ax, label="Observed Free Energy (clipped)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from mpl_toolkits.mplot3d import Axes3D

# === Load and clean data
file_path = "/home/alextu/scratch/free_energy_IRS_verkko123_wholegenome_chm13/HG00096_hap2_IR_free_energy.tsv"
df = pd.read_csv(file_path, sep="\t").drop(columns=["Sequence"])

df['Repeat'] = pd.to_numeric(df['Repeat'], errors='coerce')
df['Spacer'] = pd.to_numeric(df['Spacer'], errors='coerce')
df['Free_Energy'] = pd.to_numeric(df['Free_Energy'], errors='coerce')
df = df.dropna(subset=['Repeat', 'Spacer', 'Free_Energy'])

# === Train linear model on full data
X = sm.add_constant(df[['Repeat', 'Spacer']])
y = df['Free_Energy']
model = sm.OLS(y, X).fit()

# === Filter data for plotting only
plot_df = df[
    (df['Repeat'] > 10) & (df['Repeat'] < 200) &
    (df['Spacer'] > 10) & (df['Spacer'] < 80) &
    (df['Free_Energy'] < -45) & (df['Free_Energy'] > -100)
]

# === Group by Repeat + Spacer and compute mean Free Energy
df_grouped = (
    plot_df.groupby(['Repeat', 'Spacer'])['Free_Energy']
    .mean()
    .reset_index()
    .rename(columns={'Free_Energy': 'Mean_Free_Energy'})
)

capped_energy = np.clip(df_grouped['Mean_Free_Energy'], -100, 0)

# === Create meshgrid for regression surface (only in filtered range)
# Shrink surface dimensions for a cleaner fit
# Dynamically detect point boundaries, with optional padding
repeat_min = df_grouped['Repeat'].min()
repeat_max = df_grouped['Repeat'].max()
spacer_min = df_grouped['Spacer'].min()
spacer_max = df_grouped['Spacer'].max()

# Optional: trim off grey slab by tightening the range slightly
repeat_range = np.linspace(repeat_min, repeat_max, 50)
spacer_range = np.linspace(spacer_min, spacer_max, 50)
R, S = np.meshgrid(repeat_range, spacer_range)

grid_df = pd.DataFrame({'Repeat': R.ravel(), 'Spacer': S.ravel()})
grid_df_const = sm.add_constant(grid_df)
Z = model.predict(grid_df_const).values.reshape(R.shape)

# Mask values above -45 so they're not rendered
Z = np.where(Z > -35, np.nan, Z)

# === Plot
fig = plt.figure(figsize=(25, 6))
ax = fig.add_subplot(111, projection='3d')

# Regression surface (flat plane)
surf = ax.plot_surface(
    R, S, Z,
    cmap='Greys',
    alpha=0.4,
    edgecolor='none',
    antialiased=True
)

# Scatter of grouped mean values
obs = ax.scatter(
    df_grouped['Repeat'],
    df_grouped['Spacer'],
    df_grouped['Mean_Free_Energy'],
    c=capped_energy,
    cmap='coolwarm',
    s=40,
    edgecolors='black',
    linewidth=0.5,
    alpha=0.9
)

# Axis labels and title
ax.set_xlabel("Repeat Length", labelpad=10)
ax.set_ylabel("Spacer Length", labelpad=10)
ax.set_zlabel("Free Energy (kcal/mol)", labelpad=10)
ax.set_title("OLS Regression Surface", pad=20)

# Axis limits (zoom)
ax.set_xlim(df_grouped['Repeat'].min(), df_grouped['Repeat'].max())
ax.set_ylim(df_grouped['Spacer'].min(), df_grouped['Spacer'].max())
ax.set_zlim(-100, -45)

# Colorbar
cb = fig.colorbar(obs, ax=ax, shrink=0.6, pad=0.1)
cb.set_label("Free Energy", rotation=270, labelpad=15)

# Stats annotation from full model  move to right
coef_text = (
    f"Model: Free Energy ~ Repeat + Spacer (OLS)\n\n"
    f"Intercept: {model.params['const']:.2f}\n"
    f"Repeat Coef: {model.params['Repeat']:.2f}\n"
    f"Spacer Coef: {model.params['Spacer']:.2f}\n"
    f"R: {model.rsquared:.3f}"
)
ax.text2D(0.70, 0.98, coef_text, transform=ax.transAxes,
          fontsize=16, verticalalignment='top',
          bbox=dict(boxstyle="round,pad=0.4", facecolor="white", edgecolor="black", alpha=0.8))

plt.tight_layout()
plt.savefig('free_energy_vs_irs_OLS.pdf', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import make_pipeline
from matplotlib.colors import Normalize

# === Load and preprocess data
file_path = "/home/alextu/scratch/free_energy_IRS_verkko123_wholegenome_chm13/HG00096_hap2_IR_free_energy.tsv"
df = pd.read_csv(file_path, sep="\t").drop(columns=["Sequence"])

df['Repeat'] = pd.to_numeric(df['Repeat'], errors='coerce')
df['Spacer'] = pd.to_numeric(df['Spacer'], errors='coerce')
df['Free_Energy'] = pd.to_numeric(df['Free_Energy'], errors='coerce')
df = df.dropna(subset=['Repeat', 'Spacer', 'Free_Energy'])

# === Optional: Filter for reasonable region
df = df[
    (df['Repeat'] > 10) & (df['Repeat'] < 200) &
    (df['Spacer'] > 10) & (df['Spacer'] < 80) &
    (df['Free_Energy'] < -45) & (df['Free_Energy'] > -100)
]

# === Polynomial features
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(df[['Repeat', 'Spacer']])
y = df['Free_Energy']

# === Fit non-linear model
model = LinearRegression()
model.fit(X_poly, y)

# === Grid for surface
repeat_range = np.linspace(df['Repeat'].min(), df['Repeat'].max(), 50)
spacer_range = np.linspace(df['Spacer'].min(), df['Spacer'].max(), 50)
R, S = np.meshgrid(repeat_range, spacer_range)

# Predict surface from grid
grid_coords = np.column_stack([R.ravel(), S.ravel()])
grid_poly = poly.transform(grid_coords)
Z = model.predict(grid_poly).reshape(R.shape)

# === Group mean Free Energy for scatter
df_grouped = (
    df.groupby(['Repeat', 'Spacer'])['Free_Energy']
    .mean()
    .reset_index()
    .rename(columns={'Free_Energy': 'Mean_Free_Energy'})
)

capped_energy = np.clip(df_grouped['Mean_Free_Energy'], -100, 0)

# === Plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Surface
surf = ax.plot_surface(
    R, S, Z,
    cmap='Greys',
    alpha=0.4,
    edgecolor='none',
    antialiased=True
)

# Scatter points
obs = ax.scatter(
    df_grouped['Repeat'],
    df_grouped['Spacer'],
    df_grouped['Mean_Free_Energy'],
    c=capped_energy,
    cmap='coolwarm',
    s=40,
    edgecolors='black',
    linewidth=0.5,
    alpha=0.9
)

# Labels
ax.set_xlabel("Repeat Length", labelpad=10)
ax.set_ylabel("Spacer Length", labelpad=10)
ax.set_zlabel("Mean Free Energy (kcal/mol)", labelpad=10)
ax.set_title("Polynomial Regression Surface (Degree 2) for IR Free Energy", pad=20)

# Zoom to blue region
ax.set_xlim(df['Repeat'].min(), df['Repeat'].max())
ax.set_ylim(df['Spacer'].min(), df['Spacer'].max())
ax.set_zlim(-100, -45)

# Colorbar
cb = fig.colorbar(obs, ax=ax, shrink=0.6, pad=0.1)
cb.set_label("Mean Free Energy (clipped)", rotation=270, labelpad=15)

# Annotate R
r2 = model.score(X_poly, y)
ax.text2D(0.02, 0.98,
          f"Polynomial Regression (deg=2)\nR: {r2:.3f}",
          transform=ax.transAxes,
          fontsize=10,
          verticalalignment='top',
          bbox=dict(boxstyle="round,pad=0.4", facecolor="white", edgecolor="black", alpha=0.8))

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Load the TSV file
file_path = "/home/alextu/scratch/free_energy_IRS_verkko123_wholegenome_chm13/HG00096_hap2_IR_free_energy.tsv"
df = pd.read_csv(file_path, sep="\t").drop(columns=["Sequence"], errors='ignore')

# Convert to numeric
df['Repeat'] = pd.to_numeric(df['Repeat'], errors='coerce')
df['Spacer'] = pd.to_numeric(df['Spacer'], errors='coerce')
df['Free_Energy'] = pd.to_numeric(df['Free_Energy'], errors='coerce')
df = df.dropna(subset=['Repeat', 'Spacer', 'Free_Energy'])

# Group and compute medians
repeat_medians = df.groupby('Repeat')['Free_Energy'].median().reset_index()
spacer_medians = df.groupby('Spacer')['Free_Energy'].median().reset_index()

# Correlation
corr_repeat = stats.pearsonr(repeat_medians['Repeat'], repeat_medians['Free_Energy'])
corr_spacer = stats.pearsonr(spacer_medians['Spacer'], spacer_medians['Free_Energy'])

# Plot
sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))

# Repeat vs Free Energy
plt.subplot(1, 2, 1)
sns.regplot(data=repeat_medians, x='Repeat', y='Free_Energy', scatter=True, fit_reg=True, ci=None)
plt.title("Median Free Energy vs Repeat Length")
plt.xlabel("Repeat Length")
plt.ylabel("Median Free Energy (kcal/mol)")
plt.text(
    0.95, 0.95,
    f"r = {corr_repeat[0]:.2f}\np = {corr_repeat[1]:.3g}",
    transform=plt.gca().transAxes,
    ha='right', va='top', fontsize=12
)

# Spacer vs Free Energy
plt.subplot(1, 2, 2)
sns.regplot(data=spacer_medians, x='Spacer', y='Free_Energy', scatter=True, fit_reg=True, ci=None)
plt.title("Median Free Energy vs Spacer Length")
plt.xlabel("Spacer Length")
plt.ylabel("Median Free Energy (kcal/mol)")
plt.text(
    0.95, 0.95,
    f"r = {corr_spacer[0]:.2f}\np = {corr_spacer[1]:.3g}",
    transform=plt.gca().transAxes,
    ha='right', va='top', fontsize=12
)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = "/home/alextu/scratch/free_energy_IRS_verkko123_wholegenome/HG00096_hap1_IR_free_energy.tsv"
df = pd.read_csv(file_path, sep="\t")

# Drop heavy column
df = df.drop(columns=["Sequence"])

# Convert columns
df['Repeat'] = pd.to_numeric(df['Repeat'], errors='coerce')
df['Spacer'] = pd.to_numeric(df['Spacer'], errors='coerce')
df['Free_Energy'] = pd.to_numeric(df['Free_Energy'], errors='coerce')
df = df.dropna(subset=['Repeat', 'Spacer', 'Free_Energy'])

# Extract chromosome info from Sequence_name
df['Chromosome'] = df['Sequence_name'].str.extract(r'\|(chr[^\|]+)')

# Compute average Free Energy per chromosome
chr_avg_energy = df.groupby('Chromosome')['Free_Energy'].mean().reset_index()

# Sort by chromosome number (natural sort)
chr_avg_energy['chr_num'] = chr_avg_energy['Chromosome'].str.extract(r'chr(\d+|X|Y)').iloc[:,0]
chr_avg_energy['chr_num'] = pd.Categorical(chr_avg_energy['chr_num'],
    categories=[str(i) for i in range(1, 23)] + ['X', 'Y'], ordered=True)
chr_avg_energy = chr_avg_energy.sort_values('chr_num')

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=chr_avg_energy, x='Chromosome', y='Free_Energy', hue='Chromosome', palette='coolwarm', legend=False)
plt.title("Average IR Free Energy by Chromosome")
plt.xlabel("Chromosome")
plt.ylabel("Average Free Energy (kcal/mol)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
file_path = "/home/alextu/scratch/free_energy_IRS_verkko123_wholegenome/HG00096_hap1_IR_free_energy.tsv"
chrom = "chrY"  # <- Change this to any chromosome like "chr10", "chrX", etc.
bin_size = 500_000

# Load and clean
df = pd.read_csv(file_path, sep="\t").drop(columns=["Sequence"])
df['Start'] = pd.to_numeric(df['Start'], errors='coerce')
df['Stop'] = pd.to_numeric(df['Stop'], errors='coerce')
df['Free_Energy'] = pd.to_numeric(df['Free_Energy'], errors='coerce')
df = df.dropna(subset=['Start', 'Stop', 'Free_Energy'])

# Filter for selected chromosome
df_chr = df[df['Sequence_name'].str.contains(rf"\|{chrom}\b")].copy()

# Midpoint and binning
df_chr['Midpoint'] = (df_chr['Start'] + df_chr['Stop']) / 2
df_chr['Bin'] = (df_chr['Midpoint'] // bin_size).astype(int) * bin_size

# Average free energy per bin
bin_energy = df_chr.groupby('Bin')['Free_Energy'].mean().reset_index()

# Plot
plt.figure(figsize=(14, 5))
sns.lineplot(data=bin_energy, x='Bin', y='Free_Energy', marker='o')
plt.title(f"Average IR Free Energy Across {chrom} ({bin_size//1_000_000}Mb Bins)")
plt.xlabel("Genomic Position (bp)")
plt.ylabel("Avg Free Energy (kcal/mol)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

# All Haplotypes

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Set paths
directory = "/home/alextu/scratch/free_energy_IRS_verkko123_wholegenome/"
tsv_files = [f for f in os.listdir(directory) if f.endswith(".tsv")]

# Empty list to collect results
results = []

for file in tsv_files:
    print("Processing file", file)
    file_path = os.path.join(directory, file)
    haplotype_id = file.replace("_IR_free_energy.tsv", "")  # Extract haplotype name

    # Load file
    df = pd.read_csv(file_path, sep="\t").drop(columns=["Sequence"], errors='ignore')

    # Convert
    df['Repeat'] = pd.to_numeric(df['Repeat'], errors='coerce')
    df['Spacer'] = pd.to_numeric(df['Spacer'], errors='coerce')
    df['Free_Energy'] = pd.to_numeric(df['Free_Energy'], errors='coerce')

    # Drop NaNs
    df = df.dropna(subset=['Repeat', 'Spacer', 'Free_Energy'])

    # Group by Repeat and Spacer, average Free Energy
    repeat_means = df.groupby('Repeat')['Free_Energy'].mean().reset_index()
    spacer_means = df.groupby('Spacer')['Free_Energy'].mean().reset_index()

    # Correlations (even if small number of points)
    try:
        corr_repeat = stats.pearsonr(repeat_means['Repeat'], repeat_means['Free_Energy'])
    except Exception:
        corr_repeat = (float('nan'), float('nan'))

    try:
        corr_spacer = stats.pearsonr(spacer_means['Spacer'], spacer_means['Free_Energy'])
    except Exception:
        corr_spacer = (float('nan'), float('nan'))

    # Save results
    results.append({
        'Haplotype': haplotype_id,
        'r_repeat': corr_repeat[0],
        'p_repeat': corr_repeat[1],
        'r_spacer': corr_spacer[0],
        'p_spacer': corr_spacer[1]
    })

# Make results DataFrame
results_df = pd.DataFrame(results)

#  FIX: Create plot_df before plotting
plot_df = results_df.melt(
    id_vars='Haplotype',
    value_vars=['r_repeat', 'r_spacer'],
    var_name='Type',
    value_name='Correlation'
)

# =========================
# Plotting
# =========================
sns.set(style="whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharey=False)

# Repeat vs Free Energy
sns.boxplot(
    data=plot_df[plot_df['Type'] == 'r_repeat'],
    x='Type',
    y='Correlation',
    ax=axes[0],
    showfliers=False
)
sns.stripplot(
    data=plot_df[plot_df['Type'] == 'r_repeat'],
    x='Type',
    y='Correlation',
    ax=axes[0],
    color='black',
    alpha=0.5
)
axes[0].set_title("Repeat vs Free Energy")
axes[0].set_xlabel("")
axes[0].set_ylabel("Pearson r")
axes[0].set_xticks([])

# Spacer vs Free Energy
sns.boxplot(
    data=plot_df[plot_df['Type'] == 'r_spacer'],
    x='Type',
    y='Correlation',
    ax=axes[1],
    showfliers=False
)
sns.stripplot(
    data=plot_df[plot_df['Type'] == 'r_spacer'],
    x='Type',
    y='Correlation',
    ax=axes[1],
    color='black',
    alpha=0.5
)
axes[1].set_title("Spacer vs Free Energy")
axes[1].set_xlabel("")
axes[1].set_ylabel("Pearson r")
axes[1].set_xticks([])

plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# ==========================================
# Set up file paths
# ==========================================
directory = "/home/alextu/scratch/free_energy_IRs_verkko_123_wholegenome_hg38/"
tsv_files = [f for f in os.listdir(directory) if f.endswith(".tsv")]

# ==========================================
# Collect correlation values across haplotypes
# ==========================================
results = []

for file in tsv_files:
    print("Processing file:", file)
    file_path = os.path.join(directory, file)
    haplotype_id = file.replace("_IR_free_energy.tsv", "")

    # Load and clean data
    df = pd.read_csv(file_path, sep="\t").drop(columns=["Sequence"], errors='ignore')
    df['Repeat'] = pd.to_numeric(df['Repeat'], errors='coerce')
    df['Spacer'] = pd.to_numeric(df['Spacer'], errors='coerce')
    df['Free_Energy'] = pd.to_numeric(df['Free_Energy'], errors='coerce')
    df = df.dropna(subset=['Repeat', 'Spacer', 'Free_Energy'])

    # Compute mean-based Free Energy per bin
    repeat_means = df.groupby('Repeat')['Free_Energy'].mean().reset_index()
    spacer_means = df.groupby('Spacer')['Free_Energy'].mean().reset_index()

    # Compute correlations on means
    try:
        r_repeat, _ = stats.pearsonr(repeat_means['Repeat'], repeat_means['Free_Energy'])
    except Exception:
        r_repeat = float('nan')

    try:
        r_spacer, _ = stats.pearsonr(spacer_means['Spacer'], spacer_means['Free_Energy'])
    except Exception:
        r_spacer = float('nan')

    # Store result
    results.append({
        'Haplotype': haplotype_id,
        'r_repeat': r_repeat,
        'r_spacer': r_spacer
    })

# ==========================================
# Convert to DataFrame and save as CSV
# ==========================================
results_df = pd.DataFrame(results).dropna(subset=['r_repeat', 'r_spacer'])

# Save plotting data
results_df.to_csv("per_haplotype_correlation_data.csv", index=False)

# ==========================================
# Scatterplot: Repeat r vs Spacer r per haplotype
# ==========================================
sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))  # Rectangular: wider than tall

# Scatterplot without legend
sns.scatterplot(
    data=results_df,
    x='r_repeat',
    y='r_spacer',
    hue='r_repeat',
    palette='coolwarm',
    edgecolor='black',
    legend=False  # Disable legend
)

# Regression line
sns.regplot(
    data=results_df,
    x='r_repeat',
    y='r_spacer',
    scatter=False,
    color='black',
    line_kws={'linestyle': '--'}
)

# Axes lines
plt.axhline(0, linestyle='--', color='gray', linewidth=0.7)
plt.axvline(0, linestyle='--', color='gray', linewidth=0.7)

# Optional: Zoom
plt.xlim(-0.93, -1.0)
plt.ylim(-0.88, -0.92)

# Labels and title
plt.title("Per-Haplotype Correlation: Repeat vs Spacer Length (Mean Free Energy)")
plt.xlabel("Pearson r (Repeat Length vs Mean Free Energy)")
plt.ylabel("Pearson r (Spacer Length vs Mean Free Energy)")

plt.tight_layout()
plt.show()
```
